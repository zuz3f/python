Výsledek vypíše ve formátu : "1 + 1 = 2" atd. Na konci se Vás zeptá, zda chcete zadat jiná čísla či skončíte pomocí stisku klávesy E.

````python
#TODO
x = input("Zadej jakou operaci chceš dělat:")
print(f"Zadej jakou operaci chceš dělat: {x}")
while x != "e":
    a = int(input("Zadejte první číslo:"))
    print(f"První číslo: {a}")
    b = int(input("Zadejte druhé číslo:"))
    print(f"Druhé číslo: {b}")
    if x == "-":
        y = a - b
    print(f" {a} x {b }= {y}")
````
