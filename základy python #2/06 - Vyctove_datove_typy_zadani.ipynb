## Co jsou kolekce?

Zat√≠m jsme pracovali s promƒõnn√Ωmi, kter√© uchov√°valy **jednu hodnotu** (`int`, `str`, `bool`, `float`). V praxi ale ƒçasto pot≈ôebujeme ulo≈æit **v√≠ce hodnot** pohromadƒõ ‚Äì seznam student≈Ø, sou≈ôadnice bodu, datab√°zi kontakt≈Ø...

K tomu slou≈æ√≠ **kolekƒçn√≠ datov√© typy**:

| Typ | Syntaxe | Mƒõniteln√Ω? | Indexovan√Ω? | Duplicity? | Pou≈æit√≠ |
|:---:|:---:|:---:|:---:|:---:|---|
| `list` | `[1, 2, 3]` | ‚úÖ Ano | ‚úÖ Ano | ‚úÖ Ano | Univerz√°ln√≠ seznam |
| `tuple` | `(1, 2, 3)` | ‚ùå Ne | ‚úÖ Ano | ‚úÖ Ano | Nemƒõnn√° data (sou≈ôadnice, konstanty) |
| `set` | `{1, 2, 3}` | ‚úÖ Ano | ‚ùå Ne | ‚ùå Ne | Unik√°tn√≠ hodnoty |
| `dict` | `{"a": 1}` | ‚úÖ Ano | ‚úÖ Kl√≠ƒçem | ‚ùå Kl√≠ƒçe | P√°ry kl√≠ƒç‚Äìhodnota |
| `Enum` | `class ...` | ‚ùå Ne | ‚úÖ Jm√©nem | ‚ùå Ne | Pojmenovan√© konstanty |

> üí° **Mƒõniteln√Ω (mutable)** = po vytvo≈ôen√≠ m≈Ø≈æeme p≈ôid√°vat, odeb√≠rat a mƒõnit prvky.  
> **Nemƒõniteln√Ω (immutable)** = po vytvo≈ôen√≠ u≈æ nelze obsah zmƒõnit.

## 0Ô∏è‚É£ Rychl√© p≈ôipomenut√≠: `range()`

`range()` jsme u≈æ poznali u `for` cyklu. Technicky tak√© generuje posloupnost, ale **nevytv√°≈ô√≠ seznam v pamƒõti** ‚Äì generuje ƒç√≠sla postupnƒõ (je to tzv. *lazy* objekt):

````python
# === range() ‚Äì p≈ôipomenut√≠ ===
# range() s√°m o sobƒõ nevyp√≠≈°e ƒç√≠sla, ale vytvo≈ô√≠ objekt
print(range(5))          # range(0, 5)

# Hvƒõzdiƒçka * "rozbal√≠" range do jednotliv√Ωch hodnot
print(*range(5))         # 0 1 2 3 4

# P≈ôevod na seznam pro lep≈°√≠ vizualizaci
print(list(range(5)))    # [0, 1, 2, 3, 4]

# P≈ôipomenut√≠: range(start, stop, krok)
print(list(range(2, 10, 3)))   # [2, 5, 8]
print(list(range(10, 0, -2)))  # [10, 8, 6, 4, 2]
````

## 1Ô∏è‚É£ List (seznam) ‚Äì `[...]`

**List** je nejpou≈æ√≠vanƒõj≈°√≠ kolekce v Pythonu. Je **mƒõniteln√Ω** (mutable) a **uspo≈ô√°dan√Ω** (zachov√°v√° po≈ôad√≠ prvk≈Ø).

### Vytvo≈ôen√≠ listu

```python
prazdny = []                         # pr√°zdn√Ω list
cisla = [1, 2, 3, 4, 5]             # list ƒç√≠sel
jmena = ["Adam", "B√°ra", "Cyril"]   # list string≈Ø
mix = [1, "text", True, 3.14]       # list r≈Øzn√Ωch typ≈Ø
```

````python
# === Vytvo≈ôen√≠ listu ===
muj_list = ["jablko", "hru≈°ka", "meloun"]
print(muj_list)           # ['jablko', 'hru≈°ka', 'meloun']
print(type(muj_list))     # <class 'list'>
print(len(muj_list))      # 3 ‚Äì poƒçet prvk≈Ø
````

> ‚ö†Ô∏è **Pozor:** Pozor na jmenou konvenci! Vytvo≈ôen√≠ listu s n√°zvem "list", v√°m rozbije pr√°ci s Listy!

### Indexov√°n√≠ ‚Äì p≈ô√≠stup k prvk≈Øm

Prvky v listu jsou **indexovan√© od 0** (stejnƒõ jako u string≈Ø). K prvku p≈ôistupujeme pomoc√≠ `[index]`:

```
muj_list = ["jablko", "hru≈°ka", "meloun"]
  index:       0         1         2
z√°porn√Ω:      -3        -2        -1
```

> ‚ö†Ô∏è **Pozor:** P≈ô√≠stup k neexistuj√≠c√≠mu indexu vyhod√≠ `IndexError`!

````python
# === Indexov√°n√≠ ===
#  index:       0         1         2
muj_list = ["jablko", "hru≈°ka", "meloun"]

print(muj_list[0])    # "jablko" (prvn√≠ prvek)
print(muj_list[1])    # "hru≈°ka" (druh√Ω prvek)
print(muj_list[-1])   # "meloun" (posledn√≠ prvek)

# Slicing funguje stejnƒõ jako u string≈Ø!
print(muj_list[0:2])  # ["jablko", "hru≈°ka"] (od 0 do 1)
````

### √öprava listu ‚Äì p≈ôid√°n√≠, odebr√°n√≠, zmƒõna

Proto≈æe je list **mƒõniteln√Ω**, m≈Ø≈æeme jeho obsah libovolnƒõ upravovat:

| Operace | Metoda | P≈ô√≠klad |
|---|---|---|
| P≈ôid√°n√≠ na konec | `.append(x)` | `lst.append("nov√Ω")` |
| Vlo≈æen√≠ na pozici | `.insert(i, x)` | `lst.insert(1, "nov√Ω")` |
| Odebr√°n√≠ podle indexu | `.pop(i)` | `lst.pop(0)` |
| Odebr√°n√≠ podle hodnoty | `.remove(x)` | `lst.remove("hru≈°ka")` |
| Zmƒõna prvku | `lst[i] = x` | `lst[0] = "malina"` |
| Smaz√°n√≠ v≈°eho | `.clear()` | `lst.clear()` |

````python
# === √öprava listu ===
muj_list = ["jablko", "hru≈°ka", "meloun"]
print(f"P≈Øvodn√≠: {muj_list}")

# P≈ôid√°n√≠ na konec
muj_list.append("ban√°n")
print(f"Po append('ban√°n'): {muj_list}")

# Vlo≈æen√≠ na konkr√©tn√≠ pozici
muj_list.insert(1, "t≈ôe≈°e≈à")
print(f"Po insert(1, 't≈ôe≈°e≈à'): {muj_list}")

# Odebr√°n√≠ podle indexu (vrac√≠ odebran√Ω prvek)
odebrany = muj_list.pop(2)
print(f"Po pop(2): {muj_list} (odebr√°n: '{odebrany}')")

# Odebr√°n√≠ podle hodnoty
muj_list.remove("meloun")
print(f"Po remove('meloun'): {muj_list}")

# Zmƒõna prvku na indexu
muj_list[0] = "malina"
print(f"Po lst[0] = 'malina': {muj_list}")
````

### üìè Dal≈°√≠ u≈æiteƒçn√© metody a funkce

| Metoda / Funkce | Popis | P≈ô√≠klad |
|:---|:---|:---|
| `len(lst)` | Poƒçet prvk≈Ø | `len([1,2,3])` ‚Üí `3` |
| `sorted(lst)` | Vr√°t√≠ **nov√Ω** se≈ôazen√Ω list | `sorted([3,1,2])` ‚Üí `[1,2,3]` |
| `lst.sort()` | Se≈ôad√≠ list **na m√≠stƒõ** (zmƒõn√≠ ho) | `lst.sort()` |
| `lst.reverse()` | Obr√°t√≠ po≈ôad√≠ **na m√≠stƒõ** | `lst.reverse()` |
| `lst.index(x)` | Index prvn√≠ho v√Ωskytu `x` | `["a","b"].index("b")` ‚Üí `1` |
| `lst.count(x)` | Poƒçet v√Ωskyt≈Ø `x` | `[1,1,2].count(1)` ‚Üí `2` |
| `x in lst` | Je `x` v listu? (`True`/`False`) | `3 in [1,2,3]` ‚Üí `True` |
| `lst.copy()` | Kopie listu (mƒõlk√°) | `novy = lst.copy()` |

> ‚ö†Ô∏è **Pozor na rozd√≠l:** `sorted()` vytvo≈ô√≠ **nov√Ω** list, zat√≠mco `.sort()` zmƒõn√≠ **p≈Øvodn√≠** list a vr√°t√≠ `None`!

````python
# === U≈æiteƒçn√© metody a funkce ===
cisla = [5, 2, 8, 1, 9, 2, 5]
print(f"P≈Øvodn√≠: {cisla}")
print(f"D√©lka:   len() = {len(cisla)}")
print(f"Poƒçet 2: .count(2) = {cisla.count(2)}")
print(f"Index 8: .index(8) = {cisla.index(8)}")

# sorted() - NOV√ù list (p≈Øvodn√≠ se nezmƒõn√≠!)
serazeny = sorted(cisla)
print(f"\nsorted(cisla) = {serazeny}")
print(f"P≈Øvodn√≠ cisla = {cisla}")  # nezmƒõnƒõn!

# .sort() - zmƒõn√≠ P≈ÆVODN√ç list
cisla.sort()
print(f"\nPo cisla.sort(): {cisla}")

# .reverse() - obr√°t√≠ po≈ôad√≠
cisla.reverse()
print(f"Po .reverse():   {cisla}")

# in - kontrola p≈ô√≠tomnosti
print(f"\n5 in cisla: {5 in cisla}")
print(f"3 in cisla: {3 in cisla}")
````

### ‚ö†Ô∏è Proƒç pot≈ôebuji `.copy()`? (Reference vs. kopie)

Tohle je **jeden z nejƒçastƒõj≈°√≠ch bug≈Ø** u zaƒç√°teƒçn√≠k≈Ø. Kdy≈æ p≈ôi≈ôad√≠te list p≈ôes `=`, **nevytvo≈ô√≠te kopii** ‚Äì vytvo≈ô√≠te jen **druh√Ω odkaz (referenci)** na ten **sam√Ω** objekt v pamƒõti:

```
# ‚ùå Tohle NEKOP√çRUJE!
a = [1, 2, 3]
b = a              ‚Üê b ukazuje na STEJN√ù list jako a

‚îå‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îê
‚îÇ [1,2,3] ‚îÇ  ‚Üê jeden objekt v pamƒõti
‚îî‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îò
   ‚Üë    ‚Üë
   a    b           ‚Üê dvƒõ jm√©na, stejn√° data!
```

Kdy≈æ teƒè zmƒõn√≠te `b`, **zmƒõn√≠ se i `a`** (proto≈æe to je ten sam√Ω list):

```python
b.append(99)
print(a)   # ‚Üí [1, 2, 3, 99]  üò± a se taky zmƒõnilo!
```

### ‚úÖ ≈òe≈°en√≠: `.copy()`

```python
a = [1, 2, 3]
b = a.copy()       # ‚Üê vytvo≈ô√≠ NOV√ù nez√°visl√Ω list

‚îå‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îê     ‚îå‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îê
‚îÇ [1,2,3] ‚îÇ     ‚îÇ [1,2,3] ‚îÇ   ‚Üê dva R≈ÆZN√â objekty
‚îî‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îò     ‚îî‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îò
     ‚Üë               ‚Üë
     a               b
```

 Alternativn√≠ z√°pisy:
| Zp≈Øsob | P≈ô√≠klad |
|:---|:---|
| `.copy()` | `b = a.copy()` |
| Slicing | `b = a[:]` |
| `list()` | `b = list(a)` |

> ‚ö†Ô∏è **Pozor ‚Äì mƒõlk√° kopie (shallow copy):** `.copy()` kop√≠ruje jen **prvn√≠ √∫rove≈à**. Pokud list obsahuje vno≈ôen√© listy, ty se st√°le **sd√≠lej√≠**! Pro pln√© zkop√≠rov√°n√≠ vno≈ôen√Ωch struktur pot≈ôebujete `import copy; copy.deepcopy(a)`.

### üîÑ Proch√°zen√≠ listu cyklem `for`

List a `for` cyklus jsou **nerozluƒçn√° dvojice** ‚Äì vƒõt≈°inu pr√°ce s listy budete dƒõlat pr√°vƒõ cyklem:

```python
for prvek in muj_list:
    print(prvek)
```

Pokud pot≈ôebujete i **index** prvku, pou≈æijte funkci `enumerate()`:

```python
for index, prvek in enumerate(muj_list):
    print(f"{index}: {prvek}")
```

> üí° **Tip:** `enumerate()` je Pythonick√° alternativa k `for i in range(len(lst))`. Pou≈æ√≠vejte ji!

````python
# === Proch√°zen√≠ listu ===
ovoce = ["jablko", "hru≈°ka", "ban√°n", "t≈ôe≈°e≈à"]

# Z√°kladn√≠ for cyklus
print("--- Z√°kladn√≠ for ---")
for o in ovoce:
    print(f"  {o}")

# S indexem pomoc√≠ enumerate()
print("\n--- S enumerate() ---")
for i, o in enumerate(ovoce):
    print(f"  [{i}] {o}")

# Seƒçten√≠ ƒç√≠sel v listu
cisla = [10, 20, 30, 40]
soucet = 0
for c in cisla:
    soucet += c
print(f"\nSouƒçet {cisla} = {soucet}")
print(f"Tot√©≈æ p≈ôes sum(): {sum(cisla)}")  # vestavƒõn√° funkce!
````

### üé® Datov√© typy v listu + vno≈ôen√© listy

List m≈Ø≈æe obsahovat **libovoln√© datov√© typy** ‚Äì dokonce i **jin√© listy** (= vno≈ôen√Ω/nested list):

```
‚îå‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îê
‚îÇ  [1, "ahoj", True, [10, 20]]            ‚îÇ
‚îÇ   ‚Üë    ‚Üë      ‚Üë       ‚Üë                 ‚îÇ
‚îÇ  int  str    bool    list (vno≈ôen√Ω)      ‚îÇ
‚îî‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îò
```

> üí° **List comprehension** (zkr√°cen√Ω z√°pis): Pythonick√Ω zp≈Øsob, jak vytvo≈ôit nov√Ω list z existuj√≠c√≠ho:
> ```python
> # M√≠sto:
> druhe_mocniny = []
> for x in range(5):
>     druhe_mocniny.append(x ** 2)
>
> # Nap√≠≈°eme:
> druhe_mocniny = [x ** 2 for x in range(5)]
> ```

````python
# === R≈Øzn√© datov√© typy v listu ===
mix = [42, "Python", 3.14, True, None]
print(f"Mix: {mix}")
for prvek in mix:
    print(f"  {prvek!r:10} ‚Üí typ: {type(prvek).__name__}")

# === Vno≈ôen√© listy (2D list / matice) ===
matice = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
print(f"\nMatice: {matice}")
print(f"≈ò√°dek 0:      {matice[0]}")
print(f"Prvek [1][2]:  {matice[1][2]}")  # ≈ô√°dek 1, sloupec 2 ‚Üí 6

# === List comprehension ===
druhe_mocniny = [x ** 2 for x in range(6)]
print(f"\nDruh√© mocniny: {druhe_mocniny}")

# S podm√≠nkou
suda = [x for x in range(10) if x % 2 == 0]
print(f"Sud√° ƒç√≠sla:    {suda}")
````

## üì¶ 2. Tuple (n-tice)

**Tuple** je **nemƒõnn√°** (immutable) uspo≈ô√°dan√° kolekce. Vypad√° podobnƒõ jako list, ale pou≈æ√≠v√° **kulat√© z√°vorky** `()`:

```python
muj_tuple = (1, 2, 3)
muj_tuple = 1, 2, 3      # z√°vorky nejsou povinn√© (ale doporuƒçen√©)
prazdny = ()
jediny = (42,)            # POZOR: ƒç√°rka je nutn√° pro tuple s jedn√≠m prvkem!
```

> ‚ö†Ô∏è **Pozor na z√°vorku bez ƒç√°rky:** `(42)` je jen ƒç√≠slo `42` v z√°vork√°ch, **NE** tuple! Pro tuple s jedn√≠m prvkem mus√≠te napsat `(42,)`

### üîí Co to znamen√° ‚Äûnemƒõnn√Ω"?

Po vytvo≈ôen√≠ **nelze** mƒõnit prvky, p≈ôid√°vat ani odeb√≠rat:

```
‚úÖ Lze:   ƒçten√≠, indexov√°n√≠, slicing, for cyklus, in
‚ùå Nelze:  tuple[0] = 5, .append(), .pop(), .remove()
```

### ü§î Kdy pou≈æ√≠t tuple m√≠sto listu?

| Situace | Proƒç tuple? |
|:---|:---|
| Sou≈ôadnice `(x, y)` | Hodnoty spolu logicky pat≈ô√≠, nemƒõn√≠me je |
| Barva `(R, G, B)` | Fixn√≠ sada hodnot |
| Vr√°cen√≠ v√≠ce hodnot z funkce | `return jmeno, vek` |
| Kl√≠ƒç ve slovn√≠ku | Tuple je hashable (list ne!) |

> üí° **Tip:** Tuple je **rychlej≈°√≠** ne≈æ list a zab√≠r√° **m√©nƒõ pamƒõti**. Pokud data nepot≈ôebujete mƒõnit, pou≈æijte tuple.

````python
# === Vytvo≈ôen√≠ tuple ===
bod = (3, 7)
barva = (255, 128, 0)
jediny = (42,)   # ƒç√°rka!
bez_zavorek = 1, 2, 3

print(f"Bod:         {bod}         typ: {type(bod).__name__}")
print(f"Barva:       {barva}       typ: {type(barva).__name__}")
print(f"Jednoprvkov√Ω:{jediny}       typ: {type(jediny).__name__}")
print(f"(42) =       {(42)}          typ: {type((42)).__name__}")  # to NEN√ç tuple!

# === Indexov√°n√≠ (stejn√© jako u listu) ===
print(f"\nbod[0] = {bod[0]}")
print(f"bod[-1] = {bod[-1]}")
print(f"barva[0:2] = {barva[0:2]}")

# === NELZE mƒõnit! ===
try:
    bod[0] = 99
except TypeError as e:
    print(f"\n‚ùå bod[0] = 99 ‚Üí {e}")

try:
    bod.append(5)
except AttributeError as e:
    print(f"‚ùå bod.append(5) ‚Üí {e}")
````

### üéÅ Tuple unpacking (rozbalen√≠)

Jedn√° z neju≈æiteƒçnƒõj≈°√≠ch vlastnost√≠ tuple ‚Äì **rozbalen√≠ do promƒõnn√Ωch**:

```python
bod = (3, 7)
x, y = bod        # x = 3, y = 7
```

```
   (3, 7)
    ‚Üì  ‚Üì
    x  y
```

> üí° **Z√°mƒõna dvou promƒõnn√Ωch** v Pythonu jedn√≠m ≈ô√°dkem: `a, b = b, a` ‚Äì tohle funguje d√≠ky tuple unpacking!

````python
# === Tuple unpacking ===
souradnice = (10, 20)
x, y = souradnice
print(f"Sou≈ôadnice: {souradnice}")
print(f"x = {x}, y = {y}")

# Z√°mƒõna hodnot
a, b = 5, 10
print(f"\nP≈ôed z√°mƒõnou: a={a}, b={b}")
a, b = b, a
print(f"Po z√°mƒõnƒõ:    a={a}, b={b}")

# R≈Øzn√© datov√© typy v tuple
mix_tuple = (42, "ahoj", True, 3.14)
print(f"\nMix tuple: {mix_tuple}")
for prvek in mix_tuple:
    print(f"  {prvek!r:10} ‚Üí {type(prvek).__name__}")
````

## üî¢ 3. Set (mno≈æina)

**Set** je **neuspo≈ô√°dan√°** kolekce **unik√°tn√≠ch** prvk≈Ø. Odpov√≠d√° matematick√©mu pojmu **mno≈æina**.

```python
muj_set = {1, 2, 3}
prazdny = set()          # POZOR: {} vytvo≈ô√≠ pr√°zdn√Ω SLOVN√çK, ne set!
z_listu = set([1, 2, 2, 3])  # ‚Üí {1, 2, 3} (duplik√°ty odstranƒõny)
```

### üîë Kl√≠ƒçov√© vlastnosti

| Vlastnost | Popis |
|:---|:---|
| **Neuspo≈ô√°dan√Ω** | Prvky nemaj√≠ index ‚Äì nelze `set[0]` |
| **Bez duplik√°t≈Ø** | Ka≈æd√Ω prvek se vyskytuje **maxim√°lnƒõ jednou** |
| **Mƒõniteln√Ω** | Lze p≈ôid√°vat (`.add()`) a odeb√≠rat (`.remove()`, `.discard()`) |
| **Hashable prvky** | M≈Ø≈æe obsahovat jen nemƒõnn√© typy (`int`, `str`, `tuple` ‚Äì ale NE `list`!) |

> üí° **Kdy pou≈æ√≠t set?** Kdy≈æ pot≈ôebujete **unik√°tn√≠ hodnoty** nebo **rychl√© vyhled√°v√°n√≠** (`in` je u setu mnohem **rychlej≈°√≠** ne≈æ u listu!).

### ‚úèÔ∏è Metody pro √∫pravu

| Metoda | Popis | P≈ô√≠klad |
|:---|:---|:---|
| `.add(x)` | P≈ôid√° prvek | `s.add(5)` |
| `.remove(x)` | Odebere prvek (‚ùå KeyError pokud neexistuje) | `s.remove(5)` |
| `.discard(x)` | Odebere prvek (‚úÖ bez chyby pokud neexistuje) | `s.discard(5)` |
| `.pop()` | Odebere a vr√°t√≠ **n√°hodn√Ω** prvek | `s.pop()` |
| `.clear()` | Vypr√°zdn√≠ set | `s.clear()` |

````python
# === Vytvo≈ôen√≠ setu ===
ovoce = {"jablko", "hru≈°ka", "ban√°n"}
print(f"Set: {ovoce}")
print(f"Typ: {type(ovoce).__name__}")

# P≈ôid√°n√≠ prvku
ovoce.add("t≈ôe≈°e≈à")
print(f"\nPo add('t≈ôe≈°e≈à'):   {ovoce}")

# P≈ôid√°n√≠ duplik√°tu ‚Äì nic se nestane
ovoce.add("jablko")
print(f"Po add('jablko'):   {ovoce}")  # nezmƒõnƒõn!

# Odebr√°n√≠
ovoce.remove("hru≈°ka")
print(f"Po remove('hru≈°ka'): {ovoce}")

# discard vs remove
ovoce.discard("kiwi")  # neexistuje, ale ≈æ√°dn√° chyba
print(f"Po discard('kiwi'): {ovoce}")

try:
    ovoce.remove("kiwi")  # neexistuje ‚Üí KeyError!
except KeyError as e:
    print(f"\n‚ùå remove('kiwi') ‚Üí KeyError: {e}")
````

````python
# === Duplik√°ty jsou automaticky odstranƒõny ===
cisla = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
unikatni = set(cisla)
print(f"List:       {cisla}")
print(f"Set z listu: {unikatni}")
print(f"Poƒçet unik√°tn√≠ch: {len(unikatni)}")

# Praktick√Ω p≈ô√≠klad: odstranƒõn√≠ duplik√°t≈Ø ze seznamu
jmena = ["Anna", "Petr", "Anna", "Jana", "Petr", "Anna"]
unikatni_jmena = list(set(jmena))  # set ‚Üí zpƒõt na list
print(f"\nP≈Øvodn√≠:  {jmena}")
print(f"Unik√°tn√≠: {unikatni_jmena}")
````

### ‚öôÔ∏è Mno≈æinov√© operace

Set podporuje **matematick√© operace s mno≈æinami** ‚Äì to je jeho hlavn√≠ s√≠la!

```
    A = {1, 2, 3, 4}     B = {3, 4, 5, 6}

    Sjednocen√≠ (union):        A | B = {1, 2, 3, 4, 5, 6}
    Pr≈Ønik (intersection):     A & B = {3, 4}
    Rozd√≠l (difference):       A - B = {1, 2}
    Sym. rozd√≠l (xor):         A ^ B = {1, 2, 5, 6}
```

| Oper√°tor | Metoda | Popis |
|:---:|:---|:---|
| `A \| B` | `A.union(B)` | V≈°echny prvky z obou |
| `A & B` | `A.intersection(B)` | Jen spoleƒçn√© prvky |
| `A - B` | `A.difference(B)` | Prvky v A, kter√© **nejsou** v B |
| `A ^ B` | `A.symmetric_difference(B)` | Prvky, kter√© jsou **jen v jednom** |
| `A <= B` | `A.issubset(B)` | Je A **podmno≈æinou** B? |

````python
# === Mno≈æinov√© operace ===
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

print(f"A = {A}")
print(f"B = {B}")
print(f"A | B (sjednocen√≠): {A | B}")
print(f"A & B (pr≈Ønik):     {A & B}")
print(f"A - B (rozd√≠l):     {A - B}")
print(f"B - A (rozd√≠l):     {B - A}")
print(f"A ^ B (sym.rozd√≠l): {A ^ B}")

# Praktick√Ω p≈ô√≠klad
python_studenti = {"Anna", "Petr", "Jana", "Karel"}
web_studenti = {"Jana", "Karel", "Eva", "Tom√°≈°"}

print(f"\nStudenti Pythonu: {python_studenti}")
print(f"Studenti webu:   {web_studenti}")
print(f"Oba kurzy:       {python_studenti & web_studenti}")
print(f"Alespo≈à jeden:   {python_studenti | web_studenti}")
print(f"Jen Python:      {python_studenti - web_studenti}")
````

### üîÑ Proch√°zen√≠ setu + datov√© typy

```python
for prvek in muj_set:
    print(prvek)          # po≈ôad√≠ NEN√ç zaruƒçeno!
```

> ‚ö†Ô∏è **Pozor:** Set m≈Ø≈æe obsahovat jen **hashable** (nemƒõnn√©) typy: `int`, `float`, `str`, `bool`, `tuple`. **Nem≈Ø≈æe** obsahovat `list`, `set` ani `dict`!

````python
# === Proch√°zen√≠ setu ===
barvy = {"ƒçerven√°", "zelen√°", "modr√°"}
for barva in barvy:
    print(f"  {barva}")  # po≈ôad√≠ se m≈Ø≈æe li≈°it!

# === Datov√© typy v setu ===
mix_set = {1, "ahoj", 3.14, True, (1, 2)}  # tuple m≈Ø≈æe b√Ωt v setu
print(f"\nMix set: {mix_set}")

# List v setu NELZE:
try:
    spatny_set = {1, [2, 3]}
except TypeError as e:
    print(f"\n‚ùå List v setu: {e}")

# Rychlost vyhled√°v√°n√≠: set vs list
print(f"\n'zelen√°' in barvy:  {'zelen√°' in barvy}")
print(f"'≈ælut√°' in barvy:   {'≈ælut√°' in barvy}")
````

## üìñ 4. Dictionary / Dict (slovn√≠k)

**Slovn√≠k** ukl√°d√° data jako p√°ry **kl√≠ƒç ‚Üí hodnota** (key-value). Je to **nejpou≈æ√≠vanƒõj≈°√≠** datov√° struktura vedle listu.

```python
muj_dict = {"kl√≠ƒç1": "hodnota1", "kl√≠ƒç2": "hodnota2"}
prazdny = {}
```

```
‚îå‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îê

‚îÇ  (unik√°tn√≠)   (libovoln√Ω typ)       ‚îÇ
‚îî‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îò
```

### üîë Kl√≠ƒçov√© vlastnosti

| Vlastnost | Popis |
|:---|:---|
| **Kl√≠ƒç ‚Üí Hodnota** | P≈ô√≠stup p≈ôes kl√≠ƒç (ne p≈ôes index!) |
| **Unik√°tn√≠ kl√≠ƒçe** | Ka≈æd√Ω kl√≠ƒç se m≈Ø≈æe vyskytovat **jen jednou** |
| **Mƒõniteln√Ω** | Lze p≈ôid√°vat, mƒõnit a odeb√≠rat p√°ry |
| **Hashable kl√≠ƒçe** | Kl√≠ƒç mus√≠ b√Ωt nemƒõnn√Ω typ (`str`, `int`, `tuple`...) |
| **Libovoln√© hodnoty** | Hodnota m≈Ø≈æe b√Ωt **cokoliv** (i list, dict...) |

### ‚úèÔ∏è P≈ô√≠stup a √∫pravy

| Operace | Syntaxe | Popis |
|:---|:---|:---|
| P≈ô√≠stup | `d["kl√≠ƒç"]` | ‚ùå KeyError pokud neexistuje |
| Bezpeƒçn√Ω p≈ô√≠stup | `d.get("kl√≠ƒç", default)` | ‚úÖ Vr√°t√≠ default (nebo `None`) |
| P≈ôid√°n√≠ / Zmƒõna | `d["kl√≠ƒç"] = hodnota` | P≈ôid√° nov√Ω nebo p≈ôep√≠≈°e existuj√≠c√≠ |
| Odebr√°n√≠ | `d.pop("kl√≠ƒç")` | Odebere a vr√°t√≠ hodnotu |
| Smaz√°n√≠ | `del d["kl√≠ƒç"]` | Sma≈æe p√°r |
| V≈°echny kl√≠ƒçe | `d.keys()` | Vr√°t√≠ view kl√≠ƒç≈Ø |
| V≈°echny hodnoty | `d.values()` | Vr√°t√≠ view hodnot |
| V≈°echny p√°ry | `d.items()` | Vr√°t√≠ view (kl√≠ƒç, hodnota) |

> üí° **Tip:** V≈ædy pou≈æ√≠vejte `.get()` m√≠sto `d["kl√≠ƒç"]`, pokud si nejste 100% jist√≠, ≈æe kl√≠ƒç existuje ‚Äì vyhnete se `KeyError`!

````python
# === Vytvo≈ôen√≠ slovn√≠ku ===
student = {
    "jm√©no": "Anna",
    "vƒõk": 17,
    "≈°kola": "SP≈† Kladno",
    "pr≈Ømƒõr": 1.5
}
print(f"Slovn√≠k: {student}")
print(f"Typ:     {type(student).__name__}")

# === P≈ô√≠stup k hodnot√°m ===
print(f"\nJm√©no:  {student['jm√©no']}")
print(f"Vƒõk:    {student['vƒõk']}")

# .get() ‚Äì bezpeƒçn√Ω p≈ô√≠stup
print(f"Email:  {student.get('email', 'neuvedeno')}")  # kl√≠ƒç neexistuje ‚Üí default

# P≈ô√≠m√Ω p≈ô√≠stup na neexistuj√≠c√≠ kl√≠ƒç ‚Üí KeyError!
try:
    print(student["email"])
except KeyError as e:
    print(f"‚ùå KeyError: {e}")

# === P≈ôid√°n√≠ a zmƒõna ===
student["email"] = "anna@skola.cz"  # p≈ôid√°n√≠ nov√©ho kl√≠ƒçe
student["vƒõk"] = 18                  # zmƒõna existuj√≠c√≠ho
print(f"\nPo √∫prav√°ch: {student}")

# === Odebr√°n√≠ ===
prumer = student.pop("pr≈Ømƒõr")
print(f"Odebran√Ω pr≈Ømƒõr: {prumer}")
print(f"Po pop:     {student}")

# === Duplik√°tn√≠ kl√≠ƒçe ===
test = {"a": 1, "b": 2, "a": 99}  # druh√° "a" p≈ôep√≠≈°e prvn√≠!
print(f"\nDuplik√°tn√≠ kl√≠ƒçe: {test}")  # ‚Üí {'a': 99, 'b': 2}
````

### üîÑ Proch√°zen√≠ slovn√≠ku cyklem `for`

Slovn√≠k nab√≠z√≠ **t≈ôi zp≈Øsoby** proch√°zen√≠:

```python
# 1. Jen kl√≠ƒçe (v√Ωchoz√≠)
for klic in slovnik:

# 2. Jen hodnoty
for hodnota in slovnik.values():

# 3. Kl√≠ƒçe i hodnoty souƒçasnƒõ (nejƒçastƒõj≈°√≠!)
for klic, hodnota in slovnik.items():
```

> üí° **Tip:** `.items()` vrac√≠ tuple `(kl√≠ƒç, hodnota)` ‚Äì proto se d√° rozbalit do dvou promƒõnn√Ωch d√≠ky tuple unpacking!

````python
# === Proch√°zen√≠ slovn√≠ku ===
student = {"jm√©no": "Petr", "vƒõk": 16, "t≈ô√≠da": "IT2"}

# Jen kl√≠ƒçe
print("--- Kl√≠ƒçe ---")
for klic in student:
    print(f"  {klic}")

# Jen hodnoty
print("\n--- Hodnoty ---")
for hodnota in student.values():
    print(f"  {hodnota}")

# Kl√≠ƒçe + hodnoty (neju≈æiteƒçnƒõj≈°√≠!)
print("\n--- Kl√≠ƒçe + Hodnoty ---")
for klic, hodnota in student.items():
    print(f"  {klic}: {hodnota}")

# Kontrola existence kl√≠ƒçe
print(f"\n'jm√©no' in student: {'jm√©no' in student}")
print(f"'email' in student: {'email' in student}")
````

### üé® Datov√© typy + vno≈ôen√© slovn√≠ky

**Kl√≠ƒçe** mus√≠ b√Ωt hashable (`str`, `int`, `tuple`...), ale **hodnoty** mohou b√Ωt cokoliv ‚Äì vƒçetnƒõ list≈Ø a dal≈°√≠ch slovn√≠k≈Ø:

```python
# Slovn√≠k se seznamem jako hodnotou
predmety = {
    "IT": ["Python", "HTML", "SQL"],
    "mat": ["algebra", "geometrie"]
}

# Slovn√≠k ve slovn√≠ku (nested dict)
skola = {
    "student1": {"jm√©no": "Anna", "vƒõk": 17},
    "student2": {"jm√©no": "Petr", "vƒõk": 16}
}
```

> üí° **Dict comprehension:** Podobnƒõ jako list comprehension:
> ```python
> ctverec = {x: x**2 for x in range(5)}
> # ‚Üí {0: 0, 1: 1, 2: 4, 3: 9, 4: 16}
> ```

````python
# === Vno≈ôen√© slovn√≠ky ===
trida = {
    "Anna": {"vƒõk": 17, "pr≈Ømƒõr": 1.2, "krou≈æky": ["Python", "≈°achy"]},
    "Petr": {"vƒõk": 16, "pr≈Ømƒõr": 2.1, "krou≈æky": ["fotbal"]},
    "Jana": {"vƒõk": 17, "pr≈Ømƒõr": 1.5, "krou≈æky": ["Python", "keramika"]}
}

# P≈ô√≠stup k vno≈ôen√Ωm dat≈Øm
print(f"Anna vƒõk:     {trida['Anna']['vƒõk']}")
print(f"Petr krou≈æky: {trida['Petr']['krou≈æky']}")

# Proch√°zen√≠ vno≈ôen√©ho slovn√≠ku
print("\n--- P≈ôehled t≈ô√≠dy ---")
for jmeno, info in trida.items():
    print(f"  {jmeno}: vƒõk {info['vƒõk']}, pr≈Ømƒõr {info['pr≈Ømƒõr']}")

# Dict comprehension
ctverec = {x: x**2 for x in range(6)}
print(f"\nƒåtverce: {ctverec}")

# List nelze jako kl√≠ƒç!
try:
    spatny = {[1, 2]: "hodnota"}
except TypeError as e:
    print(f"\n‚ùå List jako kl√≠ƒç: {e}")
````

## üè∑Ô∏è 5. Enum (v√Ωƒçtov√Ω typ)

**Enum** slou≈æ√≠ k definici **pojmenovan√Ωch konstant** ‚Äì sady pevn√Ωch hodnot, ze kter√Ωch si vyb√≠r√°te. Mus√≠te ho **importovat** z modulu `enum`:

```python
from enum import Enum

class Barva(Enum):
    CERVENA = 1
    ZELENA = 2
    MODRA = 3
```

### ü§î Proƒç nepou≈æ√≠t jen promƒõnn√©?

```python
# ‚ùå Bez Enum ‚Äì chybov√©, tƒõ≈æko ƒçiteln√©
barva = "ƒçerven√°"      # p≈ôeklep ‚Üí chyba
barva = "ƒçirven√°"      # Python to neodhal√≠!

# ‚úÖ S Enum ‚Äì bezpeƒçn√©, p≈ôehledn√©
barva = Barva.CERVENA   # p≈ôeklep ‚Üí okam≈æit√° chyba
barva = Barva.CIRVENA   # AttributeError!
```

> üí° **Enum se hod√≠ pro:** dny v t√Ωdnu, smƒõry (sever/jih/...), stavy (aktivn√≠/neaktivn√≠), hern√≠ obt√≠≈ænosti, barvy...

### üìã P≈ô√≠stup k ƒçlen≈Øm Enum

| P≈ô√≠stup | P≈ô√≠klad | V√Ωsledek |
|:---|:---|:---|
| P≈ôes jm√©no | `Barva.CERVENA` | `Barva.CERVENA` |
| P≈ôes hodnotu | `Barva(1)` | `Barva.CERVENA` |
| Jm√©no ƒçlenu | `Barva.CERVENA.name` | `"CERVENA"` |
| Hodnota ƒçlenu | `Barva.CERVENA.value` | `FF0000` |

````python
from enum import Enum

# === Definice Enum ===
class Den(Enum):
    PONDELI = 1
    UTERY = 2
    STREDA = 3
    CTVRTEK = 4
    PATEK = 5
    SOBOTA = 6
    NEDELE = 7

# P≈ô√≠stup k ƒçlen≈Øm
dnes = Den.STREDA
print(f"Dnes:    {dnes}")
print(f"Jm√©no:   {dnes.name}")
print(f"Hodnota: {dnes.value}")
print(f"Typ:     {type(dnes).__name__}")

# P≈ô√≠stup p≈ôes hodnotu
print(f"\nDen(5):  {Den(5)}")
print(f"Den(5).name: {Den(5).name}")

# Porovn√°v√°n√≠
print(f"\ndnes == Den.STREDA: {dnes == Den.STREDA}")
print(f"dnes == Den.PATEK:  {dnes == Den.PATEK}")

# Enum NELZE mƒõnit!
try:
    Den.PONDELI = 99
except AttributeError as e:
    print(f"\n‚ùå Zmƒõna: {e}")
````

````python
# === Proch√°zen√≠ Enum cyklem for ===
print("V≈°echny dny:")
for den in Den:
    print(f"  {den.name:10} = {den.value}")

# Praktick√© pou≈æit√≠ v podm√≠nce
dnes = Den.SOBOTA
if dnes in (Den.SOBOTA, Den.NEDELE):
    print(f"\n{dnes.name} ‚Üí V√≠kend! üéâ")
else:
    print(f"\n{dnes.name} ‚Üí Pracovn√≠ den.")
````

### üî¢ Datov√© typy hodnot v Enum

Hodnoty v Enum mohou b√Ωt **libovoln√©ho typu** ‚Äì nejen ƒç√≠sla:

```python
class Barva(Enum):
    CERVENA = "red"       # str hodnoty
    ZELENA = "green"
    MODRA = "blue"
```

> üí° **`auto()`** ‚Äì automatick√© p≈ôi≈ôazen√≠ hodnot:
> ```python
> from enum import Enum, auto
> class Smer(Enum):
>     SEVER = auto()   # 1
>     JIH = auto()     # 2
>     VYCHOD = auto()  # 3
>     ZAPAD = auto()   # 4

````python
from enum import Enum, auto

# Enum se string hodnotami
class Barva(Enum):
    CERVENA = "red"
    ZELENA = "green"
    MODRA = "blue"

print(f"Barva.CERVENA.value = '{Barva.CERVENA.value}'")
print(f"Barva('green') = {Barva('green')}")

# Enum s auto()
class Obtiznost(Enum):
    LEHKA = auto()
    STREDNI = auto()
    TEZKA = auto()
    NIGHTMARE = auto()

print("\nObt√≠≈ænosti:")
for o in Obtiznost:
    print(f"  {o.name:12} = {o.value}")
````

### üß© Enum v kombinaci s jin√Ωmi kolekcemi

Enum se ƒçasto pou≈æ√≠v√° jako **kl√≠ƒç ve slovn√≠ku** nebo **prvek v listu/setu**:

````python
# === Enum v kolekc√≠ch ===
class Predmet(Enum):
    PYTHON = "Python"
    WEB = "Web design"
    SQL = "Datab√°ze"

# Enum jako kl√≠ƒç ve slovn√≠ku
ucebna = {
    Predmet.PYTHON: "B203",
    Predmet.WEB: "A101",
    Predmet.SQL: "B205"
}
print("Uƒçebny:")
for predmet, mistnost in ucebna.items():
    print(f"  {predmet.value:15} ‚Üí {mistnost}")

# Enum v listu
dnesni_rozvrh = [Predmet.PYTHON, Predmet.SQL, Predmet.PYTHON]
print(f"\nDne≈°n√≠ rozvrh: {[p.value for p in dnesni_rozvrh]}")

# Enum v setu
unikatni_predmety = set(dnesni_rozvrh)
print(f"Unik√°tn√≠:      {[p.value for p in unikatni_predmety]}")
````

### Kdy co pou≈æ√≠t?

| Pot≈ôebuji... | Pou≈æij |
|:---|:---|
| Upraviteln√Ω seznam prvk≈Ø | `list` |
| Nemƒõnnou skupinu hodnot (sou≈ôadnice, barvy) | `tuple` |
| Unik√°tn√≠ hodnoty, rychl√© vyhled√°v√°n√≠ | `set` |
| P√°rov√°n√≠ kl√≠ƒç ‚Üí hodnota | `dict` |
| Pojmenovan√© konstanty (stavy, smƒõry, typy) | `Enum` |

### Konverze mezi typy

```python
lst = [1, 2, 2, 3]
```
