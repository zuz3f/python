## 3ï¸âƒ£ VklÃ¡dÃ¡nÃ­ promÄ›nnÃ½ch do stringu

ÄŒasto chceme do textu vloÅ¾it **hodnotu promÄ›nnÃ©**. Python nabÃ­zÃ­ dva hlavnÃ­ zpÅ¯soby:

| ZpÅ¯sob | Syntaxe | DoporuÄenÃ­ |
|---|---|---|
| **f-string** | `f"Text {promenna}"` | âœ… ModernÃ­, pÅ™ehlednÃ½ |
| `.format()` | `"Text {}".format(promenna)` | StarÅ¡Ã­, ale stÃ¡le platnÃ½ |

### ZpÅ¯sob 1: f-string (doporuÄenÃ½ â­)

PÅ™ed uvozovky napÃ­Å¡eme pÃ­smeno `f` a do `{...}` vloÅ¾Ã­me libovolnÃ½ vÃ½raz (promÄ›nnou, vÃ½poÄet, volÃ¡nÃ­ funkce...):

````python
# === f-string â€“ modernÃ­ zpÅ¯sob ===
cislo_1 = 123
cislo_2 = 456

# JednoduchÃ½ vÃ½pis s promÄ›nnou
print(f"VÃ½sledek je: {cislo_1 + cislo_2}")

# VÃ­ce promÄ›nnÃ½ch v jednom stringu
print(f"SouÄet ÄÃ­sel {cislo_1} a {cislo_2} je {cislo_1 + cislo_2}")

# UvnitÅ™ {} mÅ¯Å¾e bÃ½t libovolnÃ½ vÃ½raz:
jmeno = "Python"
print(f"Jazyk {jmeno} mÃ¡ {len(jmeno)} znakÅ¯")
print(f"3 Ã— 7 = {3 * 7}")

# ğŸ’¡ Python automaticky pÅ™evede ÄÃ­sla na text â€“ nemusÃ­te volat str()!
````

### ZpÅ¯sob 2: `.format()` (starÅ¡Ã­)

Do textu vloÅ¾Ã­me prÃ¡zdnÃ© `{}` jako zÃ¡stupnÃ© symboly a hodnoty pÅ™edÃ¡me do funkce `.format()`:

```python

"Ahoj, {jm}!".format(jm=jmeno)     # pojmenovanÃ©
```

````python
# === .format() â€“ starÅ¡Ã­ zpÅ¯sob ===
cislo_1 = 123
cislo_2 = 456

# Varianta 1: {} se vyplÅˆujÃ­ v poÅ™adÃ­ argumentÅ¯
print("VÃ½sledek je: {}".format(cislo_1 + cislo_2))

# Varianta 2: vÃ­ce promÄ›nnÃ½ch (poÅ™adÃ­ musÃ­ odpovÃ­dat)
text = "SouÄet ÄÃ­sel {} a {} je {}".format(cislo_1, cislo_2, cislo_1 + cislo_2)
print(text)

# Varianta 3: pojmenovanÃ© zÃ¡stupnÃ© symboly (pÅ™ehlednÄ›jÅ¡Ã­)
text = "SouÄet ÄÃ­sel {a} a {b} je {vysledek}".format(
    a=cislo_1,
    b=cislo_2,
    vysledek=cislo_1 + cislo_2
)
print(text)

# ğŸ’¡ f-string je pÅ™ehlednÄ›jÅ¡Ã­ â€“ doporuÄujeme pouÅ¾Ã­vat pÅ™ednostnÄ›!
````

### ğŸ’¡ PokroÄilÃ© formÃ¡tovÃ¡nÃ­ f-stringÅ¯ â€“ Å¡Ã­Å™ka, zarovnÃ¡nÃ­, desetinnÃ¡ mÃ­sta

UvnitÅ™ `{}` ve f-stringu mÅ¯Å¾eme za dvojteÄku `:` pÅ™idat **formÃ¡tovacÃ­ instrukce**. To se hodÃ­, kdyÅ¾ chceme vytvoÅ™it **tabulku, ÃºÄtenku** nebo hezky zarovnanÃ½ vÃ½stup.

#### Syntaxe

```python
f"{hodnota:zarovnÃ¡nÃ­ Å¡Ã­Å™ka .pÅ™esnost typ}"
```

| ÄŒÃ¡st | VÃ½znam | PÅ™Ã­klad |
|---|---|---|
| `<` | Zarovnat **vlevo** | `f"{'Ahoj':<10}"` â†’ `"Ahoj      "` |
| `>` | Zarovnat **vpravo** | `f"{'Ahoj':>10}"` â†’ `"      Ahoj"` |
| `^` | Zarovnat **na stÅ™ed** | `f"{'Ahoj':^10}"` â†’ `"   Ahoj   "` |
| Å¡Ã­Å™ka (ÄÃ­slo) | MinimÃ¡lnÃ­ **poÄet znakÅ¯** | `f"{'Hi':10}"` â†’ 10 znakÅ¯ wide |
| `.2f` | DesetinnÃ© ÄÃ­slo na **2 mÃ­sta** | `f"{3.14159:.2f}"` â†’ `"3.14"` |

#### PÅ™Ã­klady

```python

cena = 3.5

# ZarovnÃ¡nÃ­ vlevo na 15 znakÅ¯ + vpravo na 8 znakÅ¯ s 1 des. mÃ­stem

# VÃ½stup: "RohlÃ­k              3.5 KÄ"
```

#### PÅ™ehled formÃ¡tovacÃ­ch znakÅ¯

| FormÃ¡t | VÃ½znam | PÅ™Ã­klad | VÃ½sledek |
|---|---|---|---|
| `:<15` | Text vlevo, Å¡Ã­Å™ka 15 | `f"{'Jablko':<15}"` | `"Jablko         "` |
| `:>10` | Text vpravo, Å¡Ã­Å™ka 10 | `f"{'Cena':>10}"` | `"      Cena"` |
| `:^20` | Text na stÅ™ed, Å¡Ã­Å™ka 20 | `f"{'Titul':^20}"` | `"       Titul        "` |
| `:.2f` | Float na 2 des. mÃ­sta | `f"{3.14159:.2f}"` | `"3.14"` |
| `:>8.1f` | Float vpravo, 8 Å¡Ã­Å™ka, 1 des. | `f"{42.567:>8.1f}"` | `"    42.6"` |
| `:05d` | CelÃ© ÄÃ­slo s nulami, Å¡Ã­Å™ka 5 | `f"{42:05d}"` | `"00042"` |

> ğŸ¯ **Tip:** ZarovnÃ¡nÃ­ se hodÃ­ zejmÃ©na pÅ™i vÃ½pisu **tabulek** a **ÃºÄtenek**, kde chcete, aby sloupce byly hezky pod sebou.

````python
# === FormÃ¡tovÃ¡nÃ­ f-stringÅ¯ â€“ Å¡Ã­Å™ka a zarovnÃ¡nÃ­ ===

# 1) ZarovnÃ¡nÃ­ textu
print(f"{'Vlevo':<15}|")      # vlevo na 15 znakÅ¯
print(f"{'Vpravo':>15}|")     # vpravo na 15 znakÅ¯
print(f"{'StÅ™ed':^15}|")      # na stÅ™ed na 15 znakÅ¯
print()

# 2) DesetinnÃ¡ mÃ­sta u float
pi = 3.14159265
print(f"Bez formÃ¡tu:  {pi}")
print(f"Na 2 mÃ­sta:   {pi:.2f}")
print(f"Na 0 mÃ­st:    {pi:.0f}")
print()

# 3) Kombinace: Å¡Ã­Å™ka + zarovnÃ¡nÃ­ + desetinnÃ¡ mÃ­sta
print("=" * 30)
print(f"{'PoloÅ¾ka':<15}{'Cena':>10}")
print("-" * 30)
print(f"{'RohlÃ­k':<15}{3.5:>10.1f}")
print(f"{'MlÃ©ko 1l':<15}{24.9:>10.1f}")
print(f"{'ÄŒokolÃ¡da':<15}{49.9:>10.1f}")
print("=" * 30)

# 4) ÄŒÃ­sla s ÃºvodnÃ­mi nulami
cislo = 42
print(f"Bez nul:   {cislo}")
print(f"S nulami:  {cislo:05d}")
````

## 4ï¸âƒ£ Metody stringÅ¯

String mÃ¡ spoustu uÅ¾iteÄnÃ½ch **metod** (funkcÃ­), kterÃ© nÃ¡m umoÅ¾ÅˆujÃ­ text upravovat. VolÃ¡me je pomocÃ­ **teÄkovÃ© notace**: `text.metoda()`.

### PÅ™ehled nejdÅ¯leÅ¾itÄ›jÅ¡Ã­ch metod

| Metoda | Co dÄ›lÃ¡ | PÅ™Ã­klad |
|---|---|---|
| `.lower()` | VÅ¡e na malÃ¡ pÃ­smena | `"Ahoj".lower()` â†’ `"ahoj"` |
| `.upper()` | VÅ¡e na velkÃ¡ pÃ­smena | `"Ahoj".upper()` â†’ `"AHOJ"` |
| `.strip()` | OdstranÃ­ mezery na krajÃ­ch | `" ahoj ".strip()` â†’ `"ahoj"` |

| `.split(sep)` | RozdÄ›lÃ­ text na seznam | `"a-b-c".split("-")` â†’ `["a","b","c"]` |
| `.find(x)` | Najde index znaku `x` | `"ahoj".find("o")` â†’ `2` |
| `.count(x)` | SpoÄÃ­tÃ¡ vÃ½skyty `x` | `"banana".count("a")` â†’ `3` |

> ğŸ’¡ **DÅ¯leÅ¾itÃ©:** Stringy jsou v Pythonu **nemÄ›nnÃ©** (immutable)! Metody **nevracejÃ­** zmÄ›nÄ›nÃ½ originÃ¡l, ale vytvÃ¡Å™Ã­ **novÃ½** string. OriginÃ¡l zÅ¯stÃ¡vÃ¡ beze zmÄ›ny.

### `.lower()` a `.upper()` â€“ zmÄ›na velikosti pÃ­smen

````python
# === .lower() a .upper() ===
text = "NÄ›JaKÃ tExT"
print(f"OriginÃ¡l:     '{text}'")
print(f".lower():     '{text.lower()}'")    # â†’ "nÄ›jakÃ½ text"
print(f".upper():     '{text.upper()}'")    # â†’ "NÄšJAKÃ TEXT"

# ğŸ’¡ PraktickÃ© pouÅ¾itÃ­: porovnÃ¡vÃ¡nÃ­ vstupu bez ohledu na velikost pÃ­smen
odpoved = input("ChceÅ¡ pokraÄovat? (ano/ne): ")
if odpoved.lower() == "ano":
    print("PokraÄujeme! âœ…")
else:
    print("KonÄÃ­me! âŒ")
````

### `.split()` â€“ rozdÄ›lenÃ­ stringu na seznam

````python
# === .split() â€“ rozdÄ›lenÃ­ textu ===
veta = "NÄ›kolik nerozdÄ›lenÃ½ch slov"

# VÃ½chozÃ­ dÄ›lenÃ­ â€“ podle mezer
slova = veta.split()
print(f"RozdÄ›leno mezerou: {slova}")  # â†’ ['NÄ›kolik', 'nerozdÄ›lenÃ½ch', 'slov']

# DÄ›lenÃ­ vlastnÃ­m znakem
text = "jablko-hruÅ¡ka-banÃ¡n"
ovoce = text.split("-")
print(f"RozdÄ›leno pomlÄkou: {ovoce}")  # â†’ ['jablko', 'hruÅ¡ka', 'banÃ¡n']

# ğŸ’¡ split() se hodÃ­ napÅ™. pro zpracovÃ¡nÃ­ CSV dat:
csv_radek = "Jan,NovÃ¡k,25,Praha"
data = csv_radek.split(",")
print(f"CSV data: {data}")
````

### `.find()` â€“ nalezenÃ­ pozice znaku

VrÃ¡tÃ­ **index** (pozici) prvnÃ­ho vÃ½skytu hledanÃ©ho znaku/textu. Pokud ho **nenajde**, vrÃ¡tÃ­ `-1`:

````python
# === .find() â€“ hledÃ¡nÃ­ v textu ===
text = "NÄ›jakÃ½ text"

# HledÃ¡me pÃ­smeno "t"
index = text.find("t")
print(f"PÃ­smeno 't' nalezeno na indexu: {index}")  # â†’ 7

# HledÃ¡me nÄ›co, co v textu nenÃ­
index_z = text.find("z")
print(f"PÃ­smeno 'z' nalezeno na indexu: {index_z}")  # â†’ -1

# HledÃ¡nÃ­ celÃ©ho slova
index_slovo = text.find("text")
print(f"Slovo 'text' zaÄÃ­nÃ¡ na indexu: {index_slovo}")  # â†’ 7

# ğŸ’¡ Pokud se znak vyskytuje vÃ­cekrÃ¡t, .find() vrÃ¡tÃ­ index PRVNÃHO vÃ½skytu
text2 = "banana"
print(f"PrvnÃ­ 'a' v 'banana': index {text2.find('a')}")  # â†’ 1
````

### `.replace()` â€“ nahrazenÃ­ textu

NahradÃ­ **vÅ¡echny vÃ½skyty** hledanÃ©ho textu jinÃ½m textem. VracÃ­ **novÃ½** string (originÃ¡l se nemÄ›nÃ­!):

### `.strip()` â€“ odstranÄ›nÃ­ mezer na krajÃ­ch

OdstranÃ­ **bÃ­lÃ© znaky** (mezery, `\n`, `\t`) ze zaÄÃ¡tku a konce stringu. Velmi uÅ¾iteÄnÃ© pÅ™i zpracovÃ¡nÃ­ uÅ¾ivatelskÃ©ho vstupu:

### `.join()` â€“ spojenÃ­ seznamu do stringu (opak `.split()`)
