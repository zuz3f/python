## ğŸ”§ ZpracovÃ¡nÃ­ dat ze souboru

### TypickÃ½ postup:

1. **PÅ™eÄÃ­st** Å™Ã¡dek
2. **`strip()`** â€“ odstranit bÃ­lÃ© znaky
3. **`split()`** â€“ rozdÄ›lit na ÄÃ¡sti (podle mezery, ÄÃ¡rky, ...)
4. **PÅ™evÃ©st** typy (`int()`, `float()`, ...)

````python
# === ZpracovÃ¡nÃ­ strukturovanÃ½ch dat ===
# Nejprve vytvoÅ™Ã­me ukÃ¡zkovÃ½ soubor:
with open("znamky.txt", "w") as f:
    f.write("Adam 1 2 1 3\n")
    f.write("BÃ¡ra 2 1 1 2\n")
    f.write("Cyril 3 2 4 2\n")
    f.write("Dana 1 1 1 1\n")

# NynÃ­ ho pÅ™eÄteme a zpracujeme:
print(f"{'JmÃ©no':<10} {'ZnÃ¡mky':<15} {'PrÅ¯mÄ›r':>6}")
print("-" * 35)

with open("znamky.txt", "r") as f:
    for radek in f:
        casti = radek.strip().split()     # ["Adam", "1", "2", "1", "3"]
        jmeno = casti[0]                   # "Adam"
        znamky = [int(z) for z in casti[1:]]  # [1, 2, 1, 3]
        prumer = sum(znamky) / len(znamky)
        print(f"{jmeno:<10} {str(znamky):<15} {prumer:>6.2f}")
````

### `split()` s vlastnÃ­m oddÄ›lovaÄem

> ğŸ’¡ Soubory s daty oddÄ›lenÃ½mi stÅ™ednÃ­kem nebo ÄÃ¡rkou se nazÃ½vajÃ­ **CSV** (Comma Separated Values) â€“ velmi bÄ›Å¾nÃ½ formÃ¡t.

### PraktickÃ¡ ukÃ¡zka â€“ parsovÃ¡nÃ­ CSV

CSV soubory jsou nejbÄ›Å¾nÄ›jÅ¡Ã­ zpÅ¯sob vÃ½mÄ›ny tabulkovÃ½ch dat. V ÄŒesku se Äasto pouÅ¾Ã­vÃ¡ **stÅ™ednÃ­k** jako oddÄ›lovaÄ (protoÅ¾e ÄÃ¡rka je u nÃ¡s desetinnÃ½ oddÄ›lovaÄ):


RuÄnÃ­ parsovÃ¡nÃ­ CSV souboru je pÅ™Ã­moÄarÃ© â€“ staÄÃ­ `split()` s pÅ™Ã­sluÅ¡nÃ½m oddÄ›lovaÄem:

````python
# === PrÃ¡ce s CSV souborem ===

# 1. VytvoÅ™Ã­me ukÃ¡zkovÃ½ CSV
with open("studenti.csv", "w", encoding="utf-8") as f:
    f.write("jmÃ©no;vÄ›k;mÄ›sto\n")
    f.write("Adam;17;Kladno\n")
    f.write("BÃ¡ra;16;Praha\n")
    f.write("Cyril;18;Beroun\n")
    f.write("Dana;17;SlanÃ½\n")

# 2. PÅ™eÄteme a zpracujeme
with open("studenti.csv", "r", encoding="utf-8") as f:
    hlavicka = f.readline().strip().split(";")  # pÅ™eskoÄÃ­me hlaviÄku
    print(f"Sloupce: {hlavicka}\n")

    for radek in f:
        jmeno, vek, mesto = radek.strip().split(";")
        vek = int(vek)
        print(f"  {jmeno} ({vek} let) â€“ bydliÅ¡tÄ›: {mesto}")
        if vek >= 18:
            print(f"    â†’ {jmeno} je plnoletÃ½/Ã¡ âœ…")
````
