## 4ï¸âƒ£ OperÃ¡tor `==` â€“ â€Je rovno"

OperÃ¡tor `==` porovnÃ¡vÃ¡ **hodnoty** dvou promÄ›nnÃ½ch. VracÃ­ `True`, pokud jsou hodnoty **stejnÃ©**.

> âš ï¸ **Pozor na zÃ¡mÄ›nu!**  
> `=` je **pÅ™iÅ™azenÃ­** (uloÅ¾enÃ­ hodnoty do promÄ›nnÃ©), `==` je **porovnÃ¡nÃ­** (zjiÅ¡tÄ›nÃ­, zda jsou hodnoty stejnÃ©).

### Co vÅ¡e lze porovnÃ¡vat?

| Typ | PÅ™Ã­klad | VÃ½sledek |

| RÅ¯znÃ© typy | `5 == "5"` | `False` âš ï¸ |
| Bool a ÄÃ­slo | `True == 1` | `True` ğŸ¤” |

> ğŸ’¡ Python porovnÃ¡vÃ¡ **obsah/hodnotu**, ne datovÃ½ typ (kromÄ› pÅ™Ã­padÅ¯ jako `5 == "5"`, kde se typy liÅ¡Ã­).

````python
# === PorovnÃ¡nÃ­ ÄÃ­sel ===
a = 2
b = 2

if a == b:
    print(f"âœ… PromÄ›nnÃ© a={a} a b={b} majÃ­ STEJNOU hodnotu")
else:
    print(f"âŒ PromÄ›nnÃ© a={a} a b={b} majÃ­ RÅ®ZNOU hodnotu")

# ğŸ§ª Zkuste zmÄ›nit b na 3 a spusÅ¥te znovu!
````

## 5ï¸âƒ£ OperÃ¡tor `is` â€“ â€Je to tentÃ½Å¾ objekt?"

OperÃ¡tor `is` je **podobnÃ½** operÃ¡toru `==`, ale funguje jinak:

| OperÃ¡tor | Co porovnÃ¡vÃ¡ | OtÃ¡zka |
|:---:|---|---|
| `==` | **Hodnotu** (obsah) | â€MajÃ­ stejnÃ½ obsah?" |
| `is` | **Identitu** (mÃ­sto v pamÄ›ti) | â€Je to tentÃ½Å¾ objekt v pamÄ›ti?" |

### Analogie ğŸ 
- `==` â†’ Dva domy majÃ­ **stejnÃ½ poÄet pokojÅ¯** (stejnÃ¡ hodnota)
- `is` â†’ Dva klÃ­Äe otevÃ­rajÃ­ **ten samÃ½ dÅ¯m** (stejnÃ½ objekt v pamÄ›ti)

### Kdy `==` a `is` dÃ¡vajÃ­ rÅ¯znÃ© vÃ½sledky?
U **seznamÅ¯** (a jinÃ½ch mÄ›nitelnÃ½ch objektÅ¯): dva seznamy mohou mÃ­t stejnÃ½ obsah, ale bÃ½t uloÅ¾eny **na rÅ¯znÃ½ch mÃ­stech v pamÄ›ti**.

> ğŸ’¡ U jednoduchÃ½ch typÅ¯ (`int`, `str`) Python optimalizuje pamÄ›Å¥ â€“ stejnÃ© hodnoty Äasto sdÃ­lÃ­ stejnÃ© mÃ­sto. Proto se `is` u ÄÃ­sel a textÅ¯ chovÃ¡ podobnÄ› jako `==`, ale **nespolÃ©hejte na to!**

````python
# === PÅ™Ã­klad: b = a (odkaz na stejnÃ½ objekt) ===
a = [1, "2", True]
b = a  # b nynÃ­ ODKAZUJE na stejnÃ½ objekt jako a

# a i b ukazujÃ­ na TOTÃ‰Å½ mÃ­sto v pamÄ›ti
print(f"a == b: {a == b}")   # True â€“ stejnÃ¡ hodnota
print(f"a is b: {a is b}")   # True â€“ stejnÃ½ objekt!
print(f"id(a): {id(a)}")     # id() vracÃ­ adresu v pamÄ›ti
print(f"id(b): {id(b)}")     # stejnÃ¡ adresa jako a

if a is b:
    print("âœ… PromÄ›nnÃ© 'a' a 'b' jsou identickÃ© (stejnÃ½ objekt)")
else:
    print("âŒ PromÄ›nnÃ© 'a' a 'b' nejsou identickÃ©")

# ğŸ’¡ ProtoÅ¾e b = a, zmÄ›na v b zmÄ›nÃ­ i a:
b.append(99)
print(f"\nPo pÅ™idÃ¡nÃ­ 99 do b:")
print(f"a = {a}")  # [1, '2', True, 99] â€“ zmÄ›nilo se i a!
````

### Dva rÅ¯znÃ© objekty se stejnou hodnotou
Pokud vytvoÅ™Ã­me dva seznamy se **stejnÃ½m obsahem** samostatnÄ›, budou na **rÅ¯znÃ½ch mÃ­stech** v pamÄ›ti:

````python
# === PÅ™Ã­klad: dva samostatnÃ© seznamy ===
a = [1, "2", True]
b = [1, "2", True]  # stejnÃ½ obsah, ale NOVÃ objekt

print(f"a == b: {a == b}")   # True â€“ stejnÃ¡ hodnota
print(f"a is b: {a is b}")   # False â€“ rÅ¯znÃ© objekty!
print(f"id(a): {id(a)}")     # rÅ¯znÃ© adresy
print(f"id(b): {id(b)}")

if a is b:
    print("âœ… PromÄ›nnÃ© 'a' a 'b' jsou identickÃ© (stejnÃ½ objekt)")
else:
    print("âŒ PromÄ›nnÃ© 'a' a 'b' NEJSOU identickÃ© (rÅ¯znÃ© objekty, ale stejnÃ½ obsah)")

# ğŸ§ª ShrnutÃ­:
# == porovnÃ¡vÃ¡ HODNOTU â†’ True (obsah je stejnÃ½)
# is porovnÃ¡vÃ¡ IDENTITU â†’ False (jsou to dva rÅ¯znÃ© objekty v pamÄ›ti)
````

## 6ï¸âƒ£ Kombinace logickÃ½ch operÃ¡torÅ¯

LogickÃ© operÃ¡tory lze samozÅ™ejmÄ› **kombinovat** v jednÃ© podmÃ­nce. Je ale dÅ¯leÅ¾itÃ© znÃ¡t **priority** (poÅ™adÃ­ vyhodnocenÃ­):

### Priorita operÃ¡torÅ¯ (od nejvyÅ¡Å¡Ã­ po nejniÅ¾Å¡Ã­)

| PoÅ™adÃ­ | OperÃ¡tor | PÅ™Ã­klad |
|:---:|:---:|---|
| 1. | `not` | `not True` â†’ `False` |
| 2. | `and` | `True and False` â†’ `False` |
| 3. | `or` | `False or True` â†’ `True` |

> âš ï¸ **DÅ¯leÅ¾itÃ© pravidlo:** `and` mÃ¡ **pÅ™ednost** pÅ™ed `or`!  
> VÃ½raz `A or B and C` se vyhodnotÃ­ jako `A or (B and C)`, nikoli jako `(A or B) and C`.

### Tip: **PouÅ¾Ã­vejte zÃ¡vorky!** ğŸ¯
I kdyÅ¾ znÃ¡te priority, zÃ¡vorky dÄ›lajÃ­ kÃ³d **ÄitelnÄ›jÅ¡Ã­m** pro vÃ¡s i ostatnÃ­:

```python

# Se zÃ¡vorkami â€“ jasnÃ©, co se dÄ›je:

```

````python
# === PÅ™Ã­klad kombinace operÃ¡torÅ¯ ===
a = 1
b = 2
soucet = a + b  # = 3

# SloÅ¾enÃ¡ podmÃ­nka se zÃ¡vorkami pro pÅ™ehlednost
# VyhodnocenÃ­: nejdÅ™Ã­ve and, pak or
if soucet == 3 and (soucet > 0 or soucet == 3) and soucet > 5:
    print("PodmÃ­nka je splnÄ›na")
else:
    print("PodmÃ­nka NENÃ splnÄ›na")

# ğŸ§ª Krok za krokem:
# soucet == 3      â†’ True
# soucet > 0       â†’ True
# soucet == 3      â†’ True
# (True or True)   â†’ True
# soucet > 5       â†’ False
# True and True and False â†’ False  â† staÄÃ­ jedno False u and!
````
