### âš ï¸ ProÄ `15.5 % 3.1` nenÃ­ pÅ™esnÄ› `3.1`? â€” NepÅ™esnost desetinnÃ½ch ÄÃ­sel

#### Jak se s tÃ­m vypoÅ™Ã¡dat?

````python
# === UkÃ¡zka nepÅ™esnosti float ===
print("0.1 + 0.2 =", 0.1 + 0.2)              # 0.30000000000000004
print("0.1 + 0.2 == 0.3?", 0.1 + 0.2 == 0.3)  # False!

# === Å˜eÅ¡enÃ­: zaokrouhlenÃ­ ===
print()
print("round(0.1 + 0.2, 1) =", round(0.1 + 0.2, 1))   # 0.3
print("round(15.5 % 3.1, 1) =", round(15.5 % 3.1, 1))  # 3.1
````
