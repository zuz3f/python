<h2>Funkce <code>type()</code></h2>
<p>Funkce <code>type()</code> nám vrátí datový typ proměnné kterou dáme do jejího argumentu.<br>
Pro vypsání výsledku funkce do konzole, použijeme funkci <code>print()</code></p>

```python
number = 34
name = "Honza"
boolean_value = False
another_number = 56.981

print(number, type(number))
print(name, type(name))
print(boolean_value, type(boolean_value))
print(another_number, type(another_number))
```

Co znamená slovo `class` zjistíme později v materiálu k objektovému programování.

<h2>Přetypování</h2>
<p>Pomocí funkcionality přetypování lze datový typ proměnné přetypovat na jiný datový typ.</p>

```python
promenna = 64 # int
promenna2 = float(promenna)
print(promenna, type(promenna), promenna2, type(promenna2))

cislo = 12
retezec = str(cislo)
zpet_cislo = int(retezec)
print(cislo, type(cislo), retezec, type(retezec), zpet_cislo, type(zpet_cislo))
```

Přetypování často používáme při načítání vstupu od uživatele. Funkce <code>input()</code> vrací vždy string a pokud chceme s hodnotou pracovat jako s číslem, musíme ji přetypovat.

```python
my_integer = int(input("Zadejte cislo:"))
print("Zadal jsi:", my_integer, type(my_integer))
```

<h2>Operátory se základními datovými typy</h2>
<p>Jak název napovídá, operátor je znak, který provede nějakou operaci.<br>Při použití operátorů je důležité s jakým <strong>datovým typem</strong> jej používáme - výsledek stejného operátoru na různém datovém typu se obykle liší a někdy ani nelze operátor použít.</p>

<h3>=</h3>
<p>Operátor <code>=</code> znamená "ulož".<br>Vezme tedy <strong>hodnotu vpravo</strong> a uloží ji do <strong>proměnné vlevo</strong>.<br>Pokud bude vpravo proměnná, použije se hodnota této proměnné, ne proměnná samotná.</p>
<p>Srovnejte s použitím znaménka = v matematice. V čem se liší?</p>

<h3>+ -</h3>
<p>Operátory <code>+</code> a <code>-</code> provádí s čísly očekávané sčítání a odčítání.<br>Co však znamená sečíst/odečíst dva řetězce? Vyzkoušejte v kódu.</p>

```python
a = 15
b = 37
scitani = a + b
odcitani = a - b
print("Cisla:", a, "+", b, "=", scitani, "|", a, "-", b, "=", odcitani)

napis1 = "Ahoj"
napis2 = "Jarda"
scitani = napis1 + napis2
print("Nápisy:",napis1,"+",napis2,"=",scitani)

odcitani = napis1 - napis2
```

<h3>*</h3>
<p>Operátor <code>*</code> provádí s čísly násobení.<br>Podobně jako v předchozím případě vynásobit dva řetězce nelze. Co nám však Python umožňuje, je vynásobit číslo a řetězec.</p>

```python
a = 15
b = 37
nasobeni = a * b
print("Cisla:", a, "+", b, "=", nasobeni)

napis = "Ahoj"
nasobeni = a * napis
print("Číslo a nápis:",a,"*",napis,"=",nasobeni)
```

<h3>/</h3>
<p>Operátor <code>/</code> provádí s čísly dělení.<br>Jeho výsledkem je vždy <strong>float</strong>, bez ohledu na typ vstupních čísel a hodnotu výsledku.</p>

```python
a = 15
b = 3
deleni = a / b
print("Integer:", a, "/", b, "=", deleni, type(deleni))

a = 15.5
b = 3.1
deleni = a / b
print("Float:", a, "/", b, "=", deleni, type(deleni))
```

<h3>// %</h3>
<p>Operátory <code>//</code> a <code>%</code> znáte z 2. třídy základní školy.<br>Předtím, než jste se dozvěděli o desetinných čísel jste dělili dvě celá čísla na "výsledek" a "zbytek". "Výsledek" je ve skutečnosti výsledek celočíselného dělení, které v Pythonu realizuje operátor <code>//</code> a "zbytku", který získáme pomocí <code>%</code> někdy říkáme <strong>modulo</strong>.</p>
<p>Možná pro vás trochu překvapivě můžeme "celočíselně se zbytkem" vydělit i datový typ float - v takovém případě bude výsledek také datového typu float.<br>Operátor <code>//</code> si pak můžete představit jako otázku "kolikrát se druhé číslo vejde do prvního?"<br>Operátor <code>%</code> jako otázku "co mi zbyde z prvního čísla, pokud odečtu druhé číslo tolikrát, dokud to půjde (ne do záporných čísel)?"</p>
<p>Proč je výsledek příkladu níže "divný" vám necháme jako procvičení hledání odpovědí na internetu.<br>Malá nápověda:</p><img src="https://www.meme-arsenal.com/memes/6f5ad8c64dac8a42cfc9960c5cbfe4a9.jpg" width="30%">

```python
a = 15
b = 3
celociselne = a // b
zbytek = a % b
print("Integer:", a, "//", b, "=", celociselne, "zbytek", zbytek, type(celociselne) ,type(zbytek) )

a = 15.5
b = 3.1
celociselne = a // b
zbytek = a % b
print("Float:", a, "//", b, "=", celociselne, "zbytek", zbytek, type(celociselne) ,type(zbytek) )
```

<h3>**</h3>
<p>Operátor <code>**</code> provádí s čísly mocnění.<br>Datový typ výsledku odpovídá datovému typu vstupu.</p>
<p>Kontrolní otázka: Co znamená např.: $ 3^{0.5} $ ?</p>

```python
a = 15
b = 3
mocnina = a ** b
print("Integer:", a, " na ", b, "=", mocnina, type(mocnina))

a = 15.5
b = 3.1
mocnina = a ** b
print("Integer:", a, " na ", b, "=", mocnina, type(mocnina))
```

<p>Operátory mohou pracovat i přímo s hodnotami (hodnoty nemusí být uložené v proměnných).</p>

```python
a = 5
nasobeni = a * 10
print("Desetinásobek",a,"je",nasobeni)
```

<h3>Operátory porovnání</h3>
<p>Tyto operátory porovnávají dvě hodnoty.<br>Jejich výsledkem je boolean (True/False)</p>
<h3>==</h3>
<p>Operátor <code>==</code> zjišťuje, zda jsou dvě hodnoty stejné</p>

```python
print(2 == 5)
print(True == False)
print("ahoj" == "ahoj")
print("jedna" == "dva")
```

<h3>< > <= >=</h3>
<p>Operátory <code><</code>, <code>></code>, <code><=</code>, <code>>=</code> zjišťují, zda je první hodnota - menší - větší - menší nebo rovna - větší nebo rovna - druhé hodnotě</p>

```python
print(2 < 5)
print(2 >= 2)
print("ahoj" < "ahoj")
print("jedna" < "dva") # proč je výsledek False?
```

<h3>Zkrácený zápis</h3>
<p>Operátory lze kombinovat tak, že nám usnadní psaní.</p>
<p>Např. pokud chceme přičíst hodnotu do proměnné, zápišeme příkaz následovně:</p>

```python
var = 0
print(var)
var = var + 2
print(var)
```

Zápis jde však zjednodušit:

```python
print(var)
var += 2 # přečte hodnotu proměnné var, přičte k hodnotě 2 a výsledek uloží zpět do proměnné var
print(var)
```

Tento zápis jde použít pro všechny matematické operace:

```python
print(var)
var -= 4 # odečte od proměnné var 4
print(var)
var *= 2 # vynásobí proměnnou dvěma
print(var)
var /= 2 # vydělí proměnnou dvěma
print(var)
var %= 2 # uloží zbytek po dělení dvěma
print(var)
```
