# 🏗️ Objektově orientované programování (OOP)

## Co je OOP a proč ho potřebujeme?

Dosud jsme programovali **procedurálně** – kód se skládal z proměnných, podmínek, cyklů a funkcí. To funguje skvěle pro menší programy, ale u složitějších projektů (hry, aplikace, weby) se kód stává nepřehledným.

**Objektově orientované programování** je přístup, kde organizujeme kód do **objektů** – samostatných jednotek, které spojují **data** (vlastnosti) a **chování** (funkce) dohromady.

### Procedurální vs. objektový přístup

```
PROCEDURÁLNÍ:                          OBJEKTOVÝ:
┌──────────────────────┐               ┌──────────────────────┐
│ hrac_jmeno = "Petr"  │               │ class Hrac:          │
│ hrac_hp = 100        │               │   jmeno = "Petr"     │
│ hrac_x = 50          │               │   hp = 100           │
│ hrac_y = 30          │               │   x = 50             │
│                      │               │   y = 30             │
│ def hrac_pohyb():    │               │   def pohyb():       │
│ def hrac_utok():     │               │   def utok():        │
│ def hrac_leceni():   │               │   def leceni():      │
│                      │               │                      │
│ nepritel_jmeno = ... │               │ class Nepritel:      │
│ nepritel_hp = ...    │               │   jmeno = ...        │
│ def nepritel_pohyb():│               │   hp = ...           │
│ ...atd...            │               │   def pohyb():       │
└──────────────────────┘               └──────────────────────┘
 ❌ Všechno roztroušené              ✅ Vše pohromadě
 ❌ Těžko přidáme 2. hráče           ✅ Snadno vytvoříme dalšího
```

> 💡 **Klíčová výhoda OOP:** Jednou napíšeme „šablonu" (třídu) a pak z ní můžeme vytvořit **libovolný počet** nezávislých objektů.

## 📐 Základní pojmy OOP

V OOP se setkáte se čtyřmi základními pojmy:

| Pojem | Vysvětlení | Příklad z reálného světa |
|:---|:---|:---|
| **Třída** (class) | „Šablona" / „stavební plán" | Výkres domu |
| **Objekt** (instance) | Konkrétní věc vytvořená podle třídy | Postavený dům |
| **Atribut** | Vlastnost objektu (data) | Barva domu, počet pater |
| **Metoda** | Chování objektu (funkce) | Otevřít dveře, rozsvítit |

```
        TŘÍDA (plán)                    OBJEKTY (konkrétní domy)
┌─────────────────────┐          ┌─────────────┐  ┌─────────────┐
│     class Dum       │          │ dum1        │  │ dum2        │
│  ─────────────────  │  ──→     │ barva: bílá │  │ barva: modrá│
│  atributy:          │  ──→     │ pater: 2    │  │ pater: 1    │
│    barva, pater     │          │             │  │             │
│  metody:            │  ──→     │ rozsvit()   │  │ rozsvit()   │
│    rozsvit()        │          │ otevri()    │  │ otevri()    │
│    otevri()         │          └─────────────┘  └─────────────┘
└─────────────────────┘          Každý dům je      Vlastní barva,
  Jeden plán pro                 nezávislý!        vlastní patra
  všechny domy
```

> 🎯 **Třída** = recept, **objekt** = hotový pokrm. Z jednoho receptu můžete uvařit mnoho pokrmů – každý s trochu jinými ingrediencemi.

## 🏁 Metoda `__init__()` – konstruktor

Když vytváříme nový objekt, Python automaticky zavolá speciální metodu `__init__()` (z angl. *initialize* = inicializovat). Říká se jí **konstruktor**.

V této metodě nastavujeme **počáteční hodnoty atributů** objektu:

```python
class Hrac:
    def __init__(self, jmeno, hp):
        self.jmeno = jmeno    # atribut 'jmeno'
        self.hp = hp           # atribut 'hp'
        self.skore = 0         # atribut 'skore' (vždy 0 na začátku)
```

```
Vytváření objektu:

  hrac1 = Hrac("Petr", 100)
            │      │     │
            ▼      ▼     ▼
  __init__(self, jmeno,  hp)
            │      │     │
            ▼      ▼     ▼
         self.jmeno = "Petr"
         self.hp = 100
         self.skore = 0
```

### Co jsou **atributy**?

Atributy jsou **proměnné, které patří konkrétnímu objektu**. Vytváříme je uvnitř `__init__()` zápisem `self.nazev = hodnota`.

| Zápis | Význam |
|:---|:---|
| `self.jmeno = jmeno` | Uloží argument `jmeno` jako atribut objektu |
| `self.skore = 0` | Vytvoří atribut s výchozí hodnotou |
| `self.pozice = [x, y]` | Atribut může být libovolného typu |

> 💡 Atributy si můžete představit jako **osobní věci** objektu – každý objekt má své vlastní.

## 🪞 Co je `self`?

Slovo `self` je **odkaz na konkrétní objekt**, se kterým právě pracujeme. Každá metoda ve třídě musí mít `self` jako **první parametr**.

### Proč `self` potřebujeme?

Představte si, že máme tři auta:

```
auto1 = Auto("Škodovka")
auto2 = Auto("Trabant")
auto3 = Auto("Ferrari")

auto2.zatrub()   ← Které auto troubí? auto2!
```

Když zavoláme `auto2.zatrub()`, Python vnitřně provede `Auto.zatrub(auto2)`. Parametr `self` tedy říká: **„Pracuj s tímto konkrétním objektem"**.

```
auto2.zatrub()
   │
   └──→  Auto.zatrub(self=auto2)
                        │
                        └──→ self.nazev  =  "Trabant"
```

| Zápis | Význam |
|:---|:---|
| `self.nazev` | Atribut **tohoto** objektu |
| `self.hp = 50` | Změní HP **tohoto** objektu (ne ostatních) |
| `def metoda(self)` | `self` = objekt, na kterém metodu voláme |

> ⚠️ **Pozor:** `self` ve třídě píšeme **vždy**, ale při volání metody ho **nepředáváme** – Python ho doplní automaticky.

```python
# Definice:
def zatrub(self):         # ← self píšeme
    print(f"{self.nazev}: Tůůů!")

# Volání:
auto2.zatrub()            # ← self NEPÍŠEME (Python doplní auto2)
```

## 🚗 Příklad: třída `Auto`

Podívejme se na kompletní příklad – třída `Auto` s atributy a metodami:


````python
# === Definice třídy Auto ===
class Auto:
    def __init__(self, nazev, x, y):
        self.nazev = nazev       # textový atribut
        self.pozice = [x, y]     # seznam [x, y]
        self.hp = 100            # výchozí hodnota

    def info(self):
        return f"Auto {self.nazev} | HP: {self.hp} | Pozice: {self.pozice}"

# === Vytvoření objektů (instancí) ===
auto1 = Auto("Škodovka", 0, 0)
auto2 = Auto("Trabant", 10, 20)
auto3 = Auto("Ferrari", 300, 400)

print(type(auto1))   # <class '__main__.Auto'>
print(auto1.info())
print(auto2.info())
print(auto3.info())
````

K atributům objektu přistupujeme pomocí **tečky** (`.`):

```python
objekt.atribut          # čtení
objekt.atribut = nová   # zápis (změna)
```


````python
# === Čtení atributů ===
print(f"Název: {auto1.nazev}")
print(f"Pozice X: {auto1.pozice[0]}")
print(f"HP: {auto1.hp}")

# === Změna atributů ===
auto1.hp = 80
print(f"\nPo poškození: HP = {auto1.hp}")

# === Každý objekt je nezávislý ===
print(f"\nauto1 HP: {auto1.hp}")    # 80 (změněno)
print(f"auto2 HP: {auto2.hp}")      # 100 (beze změny)
print(f"auto3 HP: {auto3.hp}")      # 100 (beze změny)
````

> 🎯 **Důležité:** Změna atributu u jednoho objektu **neovlivní** ostatní objekty. Každý objekt si nese **vlastní kopii** atributů.

## ⚙️ Metody – chování objektů

Metody jsou **funkce definované uvnitř třídy**. Popisují, co objekt „umí dělat".

### Rozdíl mezi funkcí a metodou

| | Funkce | Metoda |
|:---|:---|:---|
| **Kde je definovaná** | Samostatně (`def func():`) | Uvnitř třídy (`class ... def met(self):`) |
| **Jak se volá** | `funkce()` | `objekt.metoda()` |
| **Má přístup k atributům?** | Ne (potřebuje argumenty) | Ano – přes `self` |
| **Vztah k objektu** | Žádný | Patří konkrétnímu objektu |

### Typy metod

```python
class Hrac:
    def __init__(self, jmeno, hp):     # konstruktor
        self.jmeno = jmeno
        self.hp = hp

    def pozdrav(self):                  # metoda BEZ parametrů
        print(f"Ahoj, jsem {self.jmeno}!")

    def utok(self, sila):              # metoda S parametrem
        print(f"{self.jmeno} útočí silou {sila}!")

    def uzdrav(self, body):            # metoda měnící atribut
        self.hp += body
        print(f"{self.jmeno}: +{body} HP → celkem {self.hp}")
```


````python
# === Třída s různými metodami ===
class Hrac:
    def __init__(self, jmeno, hp):
        self.jmeno = jmeno
        self.hp = hp
        self.skore = 0

    def pozdrav(self):
        print(f"Ahoj, jsem {self.jmeno}!")

    def utok(self, sila):
        print(f"{self.jmeno} útočí silou {sila}!")

    def uzdrav(self, body):
        self.hp += body
        print(f"{self.jmeno}: +{body} HP → celkem {self.hp}")

    def pridej_body(self, body):
        self.skore += body

# === Použití metod ===
hrac = Hrac("Petr", 100)
hrac.pozdrav()
hrac.utok(25)
hrac.uzdrav(10)
hrac.pridej_body(50)
print(f"Skóre: {hrac.skore}")
````

### Metody vracející hodnotu (`return`)

Metody mohou také **vracet hodnotu** – stejně jako funkce:


````python
# === Metody s return ===
class Obdelnik:
    def __init__(self, a, b):
        self.a = a
        self.b = b

    def obsah(self):
        return self.a * self.b

    def obvod(self):
        return 2 * (self.a + self.b)

    def je_ctverec(self):
        return self.a == self.b

# === Použití ===
obd = Obdelnik(5, 3)
print(f"Obdélník {obd.a} × {obd.b}")
print(f"  Obsah: {obd.obsah()}")
print(f"  Obvod: {obd.obvod()}")
print(f"  Je čtverec? {obd.je_ctverec()}")

ctv = Obdelnik(4, 4)
print(f"\nObdélník {ctv.a} × {ctv.b}")
print(f"  Je čtverec? {ctv.je_ctverec()}")
````

## 🖨️ Speciální metoda `__str__()`

Když zavoláme `print()` na objekt, Python potřebuje vědět, **jak ho převést na text**. K tomu hledá speciální metodu `__str__()`.

### Co se stane BEZ `__str__()`?

Pokud třídu nevybavíme metodou `__str__()`, Python netuší, jak objekt „pěkně" zobrazit, a vypíše pouze **technickou informaci** – název třídy a adresu v paměti:

```
<__main__.Auto object at 0x000001A3F7C8B710>
                         ^^^^^^^^^^^^^^^^
                         adresa v paměti (pro nás k ničemu)
```

### Co se stane S `__str__()`?

Metoda `__str__()` říká Pythonu: **„Když se někdo zeptá, jak vypadám jako text, odpověz tímto řetězcem."**

```python
def __str__(self):
    return f"Auto {self.nazev} | HP: {self.hp}"
```

Nyní `print(auto1)` místo nečitelné adresy vypíše: `Auto Škodovka | HP: 100`

### Kdy se `__str__()` automaticky zavolá?

| Situace | Příklad |
|:---|:---|
| `print(objekt)` | `print(auto1)` |
| f-string | `f"Moje auto: {auto1}"` |
| `str(objekt)` | `str(auto1)` |
| Spojení řetězců | `"Info: " + str(auto1)` |

### Pravidla

- Metoda `__str__()` musí **vrátit řetězec** (`return str`) – **ne** tisknout (`print`)
- Nemá žádné další parametry (jen `self`)

````python
# === __str__ – jak se objekt vypíše ===
class Zvire:
    def __init__(self, druh, jmeno, vek):
        self.druh = druh
        self.jmeno = jmeno
        self.vek = vek

    def __str__(self):
        return f"{self.druh} '{self.jmeno}' ({self.vek} let)"

# === Bez __str__ bychom dostali něco jako: <__main__.Zvire object at 0x...>
z1 = Zvire("Pes", "Rex", 5)
z2 = Zvire("Kočka", "Micka", 3)

print(z1)           # volá z1.__str__()
print(z2)
print(f"Mám zvíře: {z1}")   # funguje i ve f-stringu
````

### Přehled speciálních metod (dunder methods)

Speciální metody (nazývané také **dunder methods** – z *double underscore*) mají vždy dvě podtržítka na obou stranách:

| Metoda | Kdy se volá | Příklad |
|:---|:---|:---|
| `__init__(self, ...)` | Při vytváření objektu | `obj = Trida()` |
| `__str__(self)` | Při `print(obj)` nebo `str(obj)` | `print(hrac)` |
| `__repr__(self)` | Při výpisu v konzoli, debugování | `repr(hrac)` |
| `__len__(self)` | Při `len(obj)` | `len(seznam)` |
| `__eq__(self, other)` | Při porovnání `==` | `obj1 == obj2` |

> 💡 Prozatím nám stačí `__init__()` a `__str__()`. Ostatní si ukážeme později.

## ⚠️ Častý problém: `__str__` vs `__repr__` – výpis objektu v seznamu

Když máme `__str__()`, výpis **jednoho** objektu funguje správně. Ale co když objekt dáme **do seznamu**?

```python
z1 = Zvire("Pes", "Rex", 5)
print(z1)        # ✅ Pes 'Rex' (5 let)
print([z1])      # ❌ [<__main__.Zvire object at 0x000001...>]
```

### Proč to nefunguje?

Když Python vypisuje **seznam**, nevolá na prvcích `__str__()`, ale `__repr__()`. To je jiná speciální metoda!

| Situace | Python volá |
|:---|:---|
| `print(objekt)` | `__str__()` |
| `f"text {objekt}"` | `__str__()` |
| `print([obj1, obj2])` | `__repr__()` na **každém** prvku |
| Výpis v konzoli bez `print` | `__repr__()` |

### ✅ Řešení – definovat `__repr__`

Nejjednodušší je nastavit `__repr__` tak, aby vracel totéž co `__str__`:

```python
class Zvire:
    def __init__(self, druh, jmeno, vek):
        self.druh = druh
        self.jmeno = jmeno
        self.vek = vek

    def __str__(self):
        return f"{self.druh} '{self.jmeno}' ({self.vek} let)"

    def __repr__(self):
        return self.__str__()   # vrátí totéž co __str__
```

```python
print([z1])   # ✅ [Pes 'Rex' (5 let)]
```

### 💡 Alternativa – výpis přes cyklus (bez `__repr__`)

Pokud nechcete definovat `__repr__`, můžete objekty v seznamu vypisovat **cyklem** – `print()` na jednotlivém objektu volá `__str__()`:

```python
for zvire in seznam:
    print(zvire)          # ✅ volá __str__()

# Nebo pomocí join:
print(", ".join(str(z) for z in seznam))
```

> 🎯 **Pravidlo:** Pokud plánujete dávat objekty do seznamů a vypisovat je, **vždy definujte i `__repr__`**. Nejjednodušší je `return self.__str__()`.

````python
# === Ukázka: __str__ vs __repr__ ===
class Zvire2:
    def __init__(self, druh, jmeno):
        self.druh = druh
        self.jmeno = jmeno

    def __str__(self):
        return f"{self.druh} '{self.jmeno}'"

    def __repr__(self):
        return self.__str__()    # repr vrací totéž co str

# === Jeden objekt ===
z1 = Zvire2("Pes", "Rex")
z2 = Zvire2("Kočka", "Micka")
print(z1)              # volá __str__  →  Pes 'Rex'

# === Seznam objektů ===
seznam = [z1, z2]
print(seznam)          # volá __repr__ na každém prvku  →  [Pes 'Rex', Kočka 'Micka']

# === Zkuste zakomentovat __repr__ a uvidíte rozdíl! ===
````

## 🔢 Více objektů – objekty v seznamu

Hlavní síla OOP se ukáže, když potřebujeme pracovat s **více objekty najednou**. Objekty můžeme ukládat do seznamů a procházet cyklem:


````python
# === Správa více objektů ===
class Zak:
    def __init__(self, jmeno, rocnik):
        self.jmeno = jmeno
        self.rocnik = rocnik
        self.znamky = []

    def pridej_znamku(self, znamka):
        self.znamky.append(znamka)

    def prumer(self):
        if len(self.znamky) == 0:
            return 0
        return sum(self.znamky) / len(self.znamky)

    def __str__(self):
        z = ", ".join(str(x) for x in self.znamky) if self.znamky else "žádné"
        return f"{self.jmeno} ({self.rocnik}. ročník) – známky: {z}"

# === Vytvoříme seznam žáků ===
zaci = [
    Zak("Adam", 1),
    Zak("Bára", 2),
    Zak("Cyril", 1),
    Zak("Dana", 3),
]

# === Přidáme známky ===
zaci[0].pridej_znamku(1)
zaci[0].pridej_znamku(2)
zaci[0].pridej_znamku(1)
zaci[1].pridej_znamku(3)
zaci[1].pridej_znamku(2)
zaci[2].pridej_znamku(1)
zaci[3].pridej_znamku(2)
zaci[3].pridej_znamku(2)

# === Výpis ===
print("📋 Seznam žáků:")
for zak in zaci:
    print(f"  {zak}  →  průměr: {zak.prumer():.2f}")
````

## 💡 Praktický příklad – Bankovní účet

Ukázka kompletní třídy, která kombinuje vše, co jsme se naučili:


````python
# === Třída BankovniUcet ===
class BankovniUcet:
    def __init__(self, majitel, pocatecni_zustatek=0):
        self.majitel = majitel
        self.zustatek = pocatecni_zustatek
        self.historie = []     # seznam operací

    def vloz(self, castka):
        self.zustatek += castka
        self.historie.append(f"+{castka} Kč")
        print(f"✅ Vloženo {castka} Kč → zůstatek: {self.zustatek} Kč")

    def vyber(self, castka):
        if castka > self.zustatek:
            print(f"❌ Nedostatek prostředků! (požadováno: {castka}, máte: {self.zustatek})")
            return
        self.zustatek -= castka
        self.historie.append(f"-{castka} Kč")
        print(f"✅ Vybráno {castka} Kč → zůstatek: {self.zustatek} Kč")

    def vypis_historii(self):
        print(f"\n📜 Historie účtu ({self.majitel}):")
        for operace in self.historie:
            print(f"  {operace}")
        print(f"  Aktuální zůstatek: {self.zustatek} Kč")

    def __str__(self):
        return f"Účet: {self.majitel} | Zůstatek: {self.zustatek} Kč"

# === Použití ===
ucet = BankovniUcet("Petr Novák", 1000)
print(ucet)

ucet.vloz(500)
ucet.vyber(200)
ucet.vyber(2000)      # nedostatek prostředků
ucet.vloz(100)
ucet.vypis_historii()
````

## 🔀 Více způsobů vytvoření objektu (přetěžování konstruktoru)

V jazycích jako **Java** nebo **C#** lze mít více konstruktorů se stejným názvem, ale různými parametry (**přetěžování** / *overloading*):

```java
// Java – více konstruktorů
class Bod {
    Bod(int x, int y) { ... }       // ze souřadnic
    Bod(String text)  { ... }       // z textu "3,5"
}
```

V Pythonu **to nejde** – třída může mít pouze **jeden** `__init__()`. Pokud definujete druhý, přepíše ten první.

### Jak to řešíme v Pythonu?

### 1. Výchozí hodnoty parametrů (nejčastější)

Parametrům v `__init__` dáme výchozí hodnotu – pak je volání flexibilní:

````python
# === 1. Výchozí hodnoty parametrů ===
class Bod:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def __str__(self):
        return f"Bod({self.x}, {self.y})"

# Různé způsoby vytvoření:
b1 = Bod(3, 5)       # oba argumenty
b2 = Bod(7)           # jen x, y = 0
b3 = Bod()            # oba výchozí
b4 = Bod(y=10)        # jen y, x = 0

print(b1)   # Bod(3, 5)
print(b2)   # Bod(7, 0)
print(b3)   # Bod(0, 0)
print(b4)   # Bod(0, 10)
````

### 2. `None` jako výchozí hodnota + větvení

Pokud chceme, aby `__init__` přijímal **úplně jiné kombinace** argumentů, použijeme `None` a pak rozhodujeme v kódu:


````python
# === 2. None jako výchozí hodnota ===
class Kruh:
    def __init__(self, r=None, d=None):
        if r is not None:
            self.r = r
        elif d is not None:
            self.r = d / 2
        else:
            self.r = 1    # výchozí poloměr

    def obsah(self):
        return 3.14159 * self.r ** 2

    def __str__(self):
        return f"Kruh(r={self.r})"

# === Různé způsoby vytvoření ===
k1 = Kruh(r=5)        # z poloměru
k2 = Kruh(d=10)       # z průměru → r = 5
k3 = Kruh()           # výchozí r = 1

print(k1, f"→ obsah: {k1.obsah():.2f}")
print(k2, f"→ obsah: {k2.obsah():.2f}")
print(k3, f"→ obsah: {k3.obsah():.2f}")
````

### 3. `@classmethod` – alternativní konstruktor

Pro úplně **jiný způsob** vytvoření objektu (např. z textu, ze souboru) použijeme **`@classmethod`**:

- Dekorátor `@classmethod` říká Pythonu, že metoda pracuje s **třídou** (ne s konkrétním objektem)
- Místo `self` dostane parametr `cls` (odkaz na třídu)
- Uvnitř zavoláme `cls(...)`, což vytvoří nový objekt

````python
# === 3. @classmethod – alternativní konstruktor ===
class Barva:
    def __init__(self, r, g, b):
        self.r = r
        self.g = g
        self.b = b

    @classmethod
    def z_hexkodu(cls, hex_kod):
        """Vytvoří Barvu z hex kódu, např. '#FF8800'."""
        hex_kod = hex_kod.lstrip("#")
        r = int(hex_kod[0:2], 16)
        g = int(hex_kod[2:4], 16)
        b = int(hex_kod[4:6], 16)
        return cls(r, g, b)           # cls(...) = Barva(r, g, b)

    @classmethod
    def bila(cls):
        """Vytvoří bílou barvu."""
        return cls(255, 255, 255)

    def __str__(self):
        return f"RGB({self.r}, {self.g}, {self.b})"

# === Různé způsoby vytvoření ===
b1 = Barva(255, 128, 0)           # normální __init__
b2 = Barva.z_hexkodu("#FF8800")   # alternativní konstruktor
b3 = Barva.bila()                  # pojmenovaný konstruktor

print(b1)   # RGB(255, 128, 0)
print(b2)   # RGB(255, 136, 0)
print(b3)   # RGB(255, 255, 255)
````

### Srovnání

| Přístup | Kdy použít | Příklad |
|:--------|:-----------|:--------|
| **Výchozí hodnoty** | Některé argumenty jsou volitelné | `Bod(x=0, y=0)` |
| **`None` + větvení** | Vzájemně se vylučující argumenty | `Kruh(r=5)` nebo `Kruh(d=10)` |
| **`@classmethod`** | Úplně jiný vstup (hex, text, soubor…) | `Barva.z_hexkodu("#FF8800")` |

> 💡 V Pythonu neexistuje přetěžování konstruktoru jako v Javě. Místo toho se používají **výchozí hodnoty**, **`None` s větvením** a **`@classmethod`**.

## 🤝 Interakce mezi objekty

V reálných programech (zejména hrách) objekty spolu **komunikují** – kouzelník útočí na nestvůru, lékař léčí pacienta, pošťák doručuje balík atd.

Klíčová myšlenka: **objekt může být parametrem metody jiného objektu**.

```
┌──────────────┐   zaklidlo(nestvura)   ┌──────────────┐
│  Kouzelnik   │ ─────────────────────→ │  Nestvura    │
│  mana: 30    │   kouzelník čte        │  hp: 80      │
│              │   nestvura.hp          │              │
│  mana: 20    │   a sníží ho           │  hp: 55      │
└──────────────┘                        └──────────────┘
```

### Jak to funguje?

Metoda jednoho objektu přijme **jiný objekt jako argument** a přes tečkovou notaci přistupuje k jeho atributům:

````python
# === Interakce mezi objekty ===

class Pokrm:
    def __init__(self, nazev, porce):
        self.nazev = nazev
        self.porce = porce

    def __str__(self):
        return f"{self.nazev} (zbývá {self.porce} porcí)"

class Osoba:
    def __init__(self, jmeno):
        self.jmeno = jmeno

    def snez(self, pokrm):           # ← pokrm je OBJEKT jiné třídy!
        if pokrm.porce <= 0:
            print(f"  ❌ {pokrm.nazev} už je snězený!")
            return
        pokrm.porce >= 1             # změníme atribut pokrmu
        print(f"  🍽️ {self.jmeno} snědl porci → {pokrm}")

# === Použití ===
pizza = Pokrm("Pizza", 3)
adam = Osoba("Adam")

print(pizza)           # Pizza (zbývá 3 porcí)
adam.snez(pizza)       # Adam snědl porci → Pizza (zbývá 2 porcí)
adam.snez(pizza)       # Adam snědl porci → Pizza (zbývá 1 porcí)
````

## 📋 Shrnutí

### Základní pojmy

| Pojem | Význam | Příklad |
|:---|:---|:---|
| **Třída** (`class`) | Šablona / plán pro objekty | `class Auto:` |
| **Objekt** (instance) | Konkrétní věc vytvořená z třídy | `auto1 = Auto(...)` |
| **Atribut** | Vlastnost objektu (data) | `self.nazev = "Škodovka"` |
| **Metoda** | Chování objektu (funkce) | `def jed_vlevo(self):` |
| **Konstruktor** | `__init__()` – nastaví počáteční stav | `def __init__(self, ...):` |
| `self` | Odkaz na aktuální objekt | `self.hp += 10` |

### Syntaxe třídy

```python
class NazevTridy:
    def __init__(self, arg1, arg2):
        self.atribut = arg1          # nastavení atributů

    def metoda(self):                # definice metody
        return self.atribut

    def __str__(self):               # jak se objekt vypíše
        return f"..."
```

### Vytvoření a použití objektu

```python
obj = NazevTridy("hodnota1", "hodnota2")   # vytvoření
obj.metoda()                                 # volání metody
print(obj.atribut)                           # čtení atributu
obj.atribut = "nová"                         # změna atributu
print(obj)                                   # volá __str__()
```

### Pravidla

| Pravidlo | |
|:---|:---|
| Třídy pojmenováváme **PascalCase** | `MojeTrida`, `BankovniUcet` |
| Každá metoda má `self` jako 1. parametr | `def metoda(self, ...)` |
| Atributy vytváříme v `__init__()` | `self.atribut = hodnota` |
| `self` při volání **nepíšeme** | `obj.metoda()` (ne `obj.metoda(obj)`) |
| Vždy implementujte `__str__()` | Pro přehledný výpis objektu |
| Objekty v seznamu? Definujte i `__repr__()` | `def __repr__(self): return self.__str__()` |

> 🎯 **Zapamatuj si:** Třída = šablona, objekt = konkrétní instance. Atributy = co objekt **má**, metody = co objekt **umí**. `self` = „tento konkrétní objekt".
