# 🧬 Dědičnost a polymorfismus

> **Cíl lekce:** Pochopit, jak třídy mohou **dědit** vlastnosti a metody od jiných tříd, jak **přepisovat** zděděné metody a co je **polymorfismus**.
>
> **Předpoklady:** Třídy, `__init__`, atributy, metody, `__str__`, zapouzdření, property.

---

## 🤔 Proč potřebujeme dědičnost?

Představte si, že programujete školní systém a máte třídy `Ucitel` a `Student`:

````python
# === Problém: opakující se kód ===
class Ucitel:
    def __init__(self, jmeno, vek):
        self.jmeno = jmeno
        self.vek = vek

    def predstav_se(self):
        print(f"Jsem {self.jmeno}, {self.vek} let")


class Student:
    def __init__(self, jmeno, vek):
        self.jmeno = jmeno       # duplikát
        self.vek = vek            # duplikát

    def predstav_se(self):        # duplikát
        print(f"Jsem {self.jmeno}, {self.vek} let")


u = Ucitel("Pan Novák", 45)
s = Student("Jana", 17)
u.predstav_se()
s.predstav_se()
````

> ⚠️ **Problém:** Obě třídy sdílejí stejný kód – `jmeno`, `vek`, `predstav_se()`. Při opravě chyby bychom museli opravovat na **dvou místech**.

**Řešení:** Společné vlastnosti zapíšeme **jednou** do rodičovské třídy a obě třídy od ní **zdědí**:

```
          ┌─────────────┐
          │  Osoba      │   ← rodičovská třída (předek)
          │─────────────│
          │ jmeno       │
          │ vek         │
          │─────────────│
          │ predstav se │
          └────┬────────┘
          ┌────┴─────┐
    ┌─────┴──┐  ┌────┴────┐
    │ Ucitel │  │ Student │   ← potomci (dědí od Osoba)
    │────────│  │─────────│
    │ predmet│  │ rocnik  │   ← vlastní atributy navíc
    └────────┘  └─────────┘
```

> 🎯 **Princip DRY** (Don't Repeat Yourself): Společný kód zapíšeme jednou a **zdědíme** ho.

---

## 📖 Základní pojmy

| Pojem | Význam | Příklad |
|:------|:-------|:--------|
| **Rodičovská třída** (parent / base) | Třída, od které se dědí | `Zvire` |
| **Potomek** (child / derived) | Třída, která dědí | `Pes`, `Kocka` |
| **Dědičnost** (inheritance) | Potomek přejímá atributy a metody rodiče | `Pes` má automaticky `jmeno` |
| **Přepsání** (override) | Potomek nahradí zděděnou metodu vlastní verzí | `Pes.zarvi()` místo `Zvire.zarvi()` |
| **`super()`** | Odkaz na rodičovskou třídu | `super().__init__(...)` |
| **Polymorfismus** | Stejné rozhraní, různé chování | `zarvi()` u Psa × Kočky |

> 💡 V Pythonu **každá třída** dědí od vestavěné třídy `object` – i když to nezapisujeme.

## 📐 Syntaxe dědičnosti

```
  class Potomek(Rodic):
                 │
                 └──▶  „Potomek dědí od Rodic"

  super().__init__(atribut)
  │
  └──▶  Zavolej __init__() rodičovské třídy
```

### Co potomek **automaticky získá** od rodiče:
- ✅ Všechny **metody** (pokud je nepřepíše)
- ✅ Všechny **atributy** nastavené v `__init__` rodiče (pokud zavolá `super().__init__()`)
- ✅ Metodu `__str__` (pokud ji nepřepíše)

````python
# === Syntaxe dědičnosti ===
class Rodic:
    def __init__(self, atribut):
        self.atribut = atribut

    def metoda(self):
        print(f"Rodičovská metoda, atribut = {self.atribut}")


class Potomek(Rodic):                     # ← závorky = dědí od Rodic
    def __init__(self, atribut, novy):
        super().__init__(atribut)          # ← zavolá __init__ rodiče
        self.novy = novy

# === Test ===
p = Potomek("ahoj", 42)
print(p.atribut)     # ahoj   ← zděděný atribut
print(p.novy)        # 42     ← vlastní atribut
p.metoda()           # Rodičovská metoda, atribut = ahoj  ← zděděná metoda
````

## 🐾 Příklad – Rodičovská třída `Zvire`

Začneme rodičovskou třídou, která obsahuje **společné vlastnosti** všech zvířat:

````python
# === Rodičovská třída Zvire ===
class Zvire:
    def __init__(self, jmeno, hmotnost):
        if hmotnost < 0:
            raise ValueError("Hmotnost nemůže být záporná")
        self.jmeno = jmeno
        self.hmotnost = hmotnost

    def zarvi(self):
        print("Huááá!")

    def __str__(self):
        return f"{self.jmeno} ({self.hmotnost} kg)"

# === Test ===
z = Zvire("Anička", 30)
print(z)           # Anička (30 kg)
z.zarvi()          # Huááá!
````

## 🐕 Potomci – `Pes` a `Kocka`

Pes i kočka jsou zvířata, ale každý **řve jinak** a má **vlastní atribut navíc**:

````python
# === Potomci ===
class Pes(Zvire):
    def __init__(self, jmeno, hmotnost, hlasitost):
        super().__init__(jmeno, hmotnost)   # zavolá Zvire.__init__
        self.hlasitost = hlasitost          # nový atribut

    def zarvi(self):                        # přepsání (override)
        print(f"Haf haf! (hlasitost {self.hlasitost})")

    def __str__(self):
        return f"Pes {super().__str__()}, hlasitost {self.hlasitost}"


class Kocka(Zvire):
    def __init__(self, jmeno, hmotnost, roztomilost):
        super().__init__(jmeno, hmotnost)
        self.roztomilost = roztomilost

    def zarvi(self):
        print(f"Mňau! (roztomilost {self.roztomilost})")

    def __str__(self):
        return f"Kočka {super().__str__()}, roztomilost {self.roztomilost}"


# === Test ===
p = Pes("Bára", 40, 90)
k = Kocka("Cecilka", 3, 100)

print(p)            # Pes Bára (40 kg), hlasitost 90
p.zarvi()           # Haf haf! (hlasitost 90)

print(k)            # Kočka Cecilka (3 kg), roztomilost 100
k.zarvi()           # Mňau! (roztomilost 100)
````

### 🔍 Co se děje krok za krokem

```
  p = Pes("Bára", 40, 90)
  │
  ├──▶ Pes.__init__ se spustí
  │       │
  │       ├──▶ super().__init__("Bára", 40)
  │       │       │
  │       │       ├──▶ self.jmeno = "Bára"
  │       │       └──▶ self.hmotnost = 40
  │       │
  │       └──▶ self.hlasitost = 90
  │
  └──▶ Výsledek: objekt Pes s atributy jmeno, hmotnost, hlasitost
```

> ⚠️ **Bez `super().__init__()`** by potomek **neměl** atributy rodiče (`jmeno`, `hmotnost`).

## 🔑 `super()` – volání rodičovské třídy

`super()` vrací odkaz na **rodičovskou třídu**. Nejčastěji ho používáme v `__init__()`.

### Dva způsoby zápisu:

| Způsob | Zápis | Doporučeno |
|:-------|:------|:-----------|
| **`super()`** | `super().__init__(jmeno, hmotnost)` | ✅ Ano |
| **Přímé volání** | `Zvire.__init__(self, jmeno, hmotnost)` | ⚠️ Funguje, ale méně flexibilní |

> 🎯 **Používejte `super()`** – je kratší, čitelnější a správně funguje i při víceúrovňové dědičnosti.

````python
# === super() v metodě __str__ ===
class Pes(Zvire):
    def __init__(self, jmeno, hmotnost, hlasitost):
        super().__init__(jmeno, hmotnost)
        self.hlasitost = hlasitost

    def __str__(self):
        # super().__str__() vrátí "Bára (40 kg)" od rodiče
        zaklad = super().__str__()
        return f"Pes {zaklad}, hlasitost {self.hlasitost}"

# === Test ===
p = Pes("Bára", 40, 90)
print(p)             # Pes Bára (40 kg), hlasitost 90
print(super(Pes, p).__str__())   # Bára (40 kg)  ← jen rodičovský __str__
````

## ✏️ Přepisování metod (override)

Potomek může **přepsat** (nahradit) libovolnou metodu rodiče tím, že definuje metodu **se stejným názvem**.

```
  Zvire.zarvi()  →  "Huááá!"
       │
       │  Pes PŘEPÍŠE zarvi()
       ▼
  Pes.zarvi()    →  "Haf haf!"
```

Existují dva přístupy:

````python
# === 1. Úplné přepsání – kompletně nový kód ===
class PesUplne(Zvire):
    def __init__(self, jmeno, hmotnost):
        super().__init__(jmeno, hmotnost)

    def zarvi(self):
        print("Haf haf!")          # kompletně nahrazeno

p1 = PesUplne("Rex", 25)
p1.zarvi()    # Haf haf!  (rodičovské "Huááá!" se nevolá)
````

````python
# === 2. Rozšíření – rodič + vlastní kód ===
class PesRozsireny(Zvire):
    def __init__(self, jmeno, hmotnost):
        super().__init__(jmeno, hmotnost)

    def zarvi(self):
        super().zarvi()            # nejprve rodičovské: "Huááá!"
        print("...a pak Haf!")     # potom vlastní

p2 = PesRozsireny("Aria", 30)
p2.zarvi()
# Huááá!
# ...a pak Haf!
````

> 💡 **Rozšíření** je užitečné, když chceme zachovat chování rodiče a jen k němu něco přidat.

### 🔄 Co se stane, když metodu nepřepíšeme?

Pokud potomek metodu **nedefinuje**, automaticky se použije verze od rodiče:

````python
# === Dědění bez přepsání ===
class Papousek(Zvire):
    # Papoušek nemá žádné vlastní metody – vše dědí
    pass

# === Test ===
pp = Papousek("Pepa", 1)
print(pp)          # Pepa (1 kg)       ← zděděný __str__
pp.zarvi()         # Huááá!            ← zděděná metoda
````

## 🔎 `isinstance()` a `issubclass()`

Python nabízí funkce pro kontrolu dědičnosti za běhu programu:

| Funkce | Co zjistí | Příklad |
|:-------|:----------|:--------|
| `isinstance(obj, Třída)` | Je objekt instancí třídy (nebo jejího předka)? | `isinstance(pes, Zvire)` → `True` |
| `issubclass(A, B)` | Dědí třída A od třídy B? | `issubclass(Pes, Zvire)` → `True` |

````python
# === isinstance a issubclass ===
p = Pes("Bára", 40, 90)
k = Kocka("Cecilka", 3, 100)

# isinstance – kontrola objektu
print(isinstance(p, Pes))       # True  – p je Pes
print(isinstance(p, Zvire))     # True  – Pes JE Zvíře (dědí)
print(isinstance(p, Kocka))     # False – Pes NENÍ Kočka

print()

# issubclass – kontrola třídy
print(issubclass(Pes, Zvire))   # True
print(issubclass(Zvire, Pes))   # False – rodič NENÍ potomkem
````

## 🎭 Polymorfismus (mnohotvarost)

**Polymorfismus** znamená, že **stejné rozhraní** (název metody) se **chová jinak** podle typu objektu.

V našem příkladu: metoda `zarvi()` existuje u `Zvire`, `Pes` i `Kocka` – ale každá verze dělá něco jiného:

```
  zarvi()
  │
  ├── u Zvire  →  "Huááá!"
  ├── u Pes    →  "Haf haf!"
  └── u Kocka  →  "Mňau!"
```

### Proč je to užitečné?

Můžeme napsat funkci, která pracuje s **libovolným zvířetem**, a ona se automaticky přizpůsobí:

````python
# === Polymorfismus v praxi ===
def veterinarni_osetreni(zvire):
    print(f"Ošetřuji: {zvire}")
    zvire.zarvi()               # každé zvíře řve jinak!
    print("Hotovo.\n")

# === Test ===
z = Zvire("Anička", 30)
p = Pes("Bára", 40, 90)
k = Kocka("Cecilka", 3, 100)

veterinarni_osetreni(z)
veterinarni_osetreni(p)
veterinarni_osetreni(k)
````

> 💡 Funkce `veterinarni_osetreni()` **neví**, jestli dostane Psa, Kočku nebo jiné zvíře. Přesto funguje správně, protože každé zvíře má metodu `zarvi()` – jen každé ji provádí **jinak**.

### Polymorfismus se seznamem

Nejčastěji polymorfismus využijeme při procházení **seznamu objektů různých typů**:

````python
# === Polymorfismus se seznamem ===
zvirata = [
    Pes("Rex", 35, 80),
    Kocka("Míca", 4, 95),
    Zvire("Had", 2),
]

for z in zvirata:
    print(z)
    z.zarvi()
    print()
````

## ⚖️ Kdy (ne)použít dědičnost

### Test „JE" vs. „MÁ"

| Otázka | Vztah | Řešení |
|:-------|:------|:-------|
| Pes **JE** Zvíře? | ✅ **Dědičnost** | `class Pes(Zvire)` |
| Auto **MÁ** Volant? | ❌ **Kompozice** | `self.volant = Volant()` |

````python
# === JE vs. MÁ – ukázka ===

# ✅ SPRÁVNĚ – Pes JE Zvíře → dědičnost
class Pes(Zvire):
    pass

# ❌ ŠPATNĚ – Auto NENÍ Volant → nemá dědit
# class Auto(Volant):    # Auto JE Volant?? NE!
#     pass

# ✅ SPRÁVNĚ – Auto MÁ Volant → kompozice (atribut)
class Volant:
    def otoc(self, smer):
        print(f"Otáčím {smer}")

class Auto:
    def __init__(self):
        self.volant = Volant()    # Auto OBSAHUJE Volant

a = Auto()
a.volant.otoc("vlevo")
````

> 🎯 **Pravidlo:** Dědičnost = vztah **„JE"** (is-a). Pokud jde o vztah **„MÁ"** (has-a), použijte atribut (kompozici).

## 🔄 Před a po – srovnání

| | Bez dědičnosti | S dědičností |
|:---|:---|:---|
| **Opakování kódu** | ❌ Ano (3× stejný kód) | ✅ Ne |
| **Oprava chyby** | Opravit na 3 místech | Opravit jednou |
| **Nový typ** | Celý kód znovu | Jen nové atributy |
| **Přehlednost** | ❌ Obtížně čitelné | ✅ Jasná hierarchie |

````python
# === Před (BEZ dědičnosti) ===
class Ucitel_stary:
    def __init__(self, jmeno, vek):
        self.jmeno = jmeno
        self.vek = vek
    def predstav_se(self):
        print(f"Jsem {self.jmeno}")

class Student_stary:
    def __init__(self, jmeno, vek):
        self.jmeno = jmeno      # duplikát
        self.vek = vek           # duplikát
    def predstav_se(self):       # duplikát
        print(f"Jsem {self.jmeno}")


# === Po (S dědičností) ===
class Osoba:
    def __init__(self, jmeno, vek):
        self.jmeno = jmeno
        self.vek = vek
    def predstav_se(self):
        print(f"Jsem {self.jmeno}, {self.vek} let")

class Ucitel(Osoba):
    def __init__(self, jmeno, vek, predmet):
        super().__init__(jmeno, vek)
        self.predmet = predmet

class Student(Osoba):
    def __init__(self, jmeno, vek, rocnik):
        super().__init__(jmeno, vek)
        self.rocnik = rocnik

# === Test ===
u = Ucitel("Pan Novák", 45, "Python")
s = Student("Jana", 17, 2)
u.predstav_se()
s.predstav_se()
````

## Cvičení 2 – Vozidla

Vytvořte rodičovskou třídu `Vozidlo` a dva potomky: `Auto` a `Motorka`.

**`Vozidlo`:**
- Atributy: `znacka`, `rychlost` (aktuální, začíná na 0)
- Metoda `zrychleni(o)` – zvýší rychlost o zadanou hodnotu

## 📋 Shrnutí

| Pojem | Popis |
|:------|:------|
| **Dědičnost** | Potomek přejímá atributy a metody rodiče |
| **`class Potomek(Rodic)`** | Syntaxe pro dědičnost |
| **`super().__init__()`** | Zavolá `__init__` rodiče |
| **Override (přepsání)** | Potomek definuje metodu se stejným názvem |
| **Rozšíření** | `super().metoda()` + vlastní kód |
| **Polymorfismus** | Stejný název metody, různé chování podle typu |
| **`isinstance(obj, Třída)`** | Kontrola, zda objekt patří do třídy (nebo potomka) |
| **JE → dědičnost** | `Pes JE Zvíře` → `class Pes(Zvire)` |
| **MÁ → kompozice** | `Auto MÁ Motor` → `self.motor = Motor()` |

> 🎯 **Na co nezapomenout:**
> - Vždy volejte `super().__init__()` v potomkovi
> - Dědičnost = vztah **JE** (is-a)
> - Kompozice = vztah **MÁ** (has-a)
> - Polymorfismus = stejné rozhraní, různé chování
